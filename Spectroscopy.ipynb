### 1.1. Key concepts of Spectroscopy

The whole electromagnetic spectrum covers the radiation in very broad range from gamma rays (emitted by the nuclei of certain radioactive elements), X-rays (used for medical examination of bones), to ultraviolet (UV) light (is responsible for sunburn, can also be used for dis-infection purpose), visible light, microwaves, and radio-frequency waves (used for radio and television communication, and of the cell phone signal).

<a href="https://en.wikipedia.org/wiki/Spectroscopy" target="_blank">Spectroscopy</a> is the study of how matter interacts with electromagnetic radiation. It is a broad scientific field that includes many sub-disciplines, each focusing on different aspects of matter and energy interactions.

The types of spectroscopy are classified based on the following criteria:

- *Type of Radiative Energy* Involved. Spectroscopy can be categorized based on the type of electromagnetic radiation used in the interaction. Common classifications include microwave, terahertz, infrared, near-infrared, ultraviolet-visible, X-ray, and gamma-ray spectroscopy, each corresponding to a specific wavelength region of the electromagnetic spectrum.

- *Nature of the Interaction*. Spectroscopy can also be distinguished by how the radiation interacts with the material. Some key types of interactions include: 
    - Absorption - measures how a substance absorbs specific wavelengths of radiation, helping identify its chemical composition; 
    - Emission - studies the light emitted by excited atoms or molecules as they return to a lower energy state; 
    - Fluorescence - examines the emission of light by a substance after it absorbs higher-energy radiation; 
    - Elastic scattering and Reflection spectroscopy - determine how incident radiation is reflected or scattered by a material.
    - Inelastic scattering - involves an exchange of energy between the radiation and the matter, shifting the wavelength of the scattered radiation (Raman and Compton scattering).

- *Type of Material*. Different spectroscopic techniques are designed to study specific types of matter, including atoms, molecules, crystals, gases, liquids, solids, and plasmas. Each requires specialized methods to extract structural and compositional information. For example, atomic spectroscopy focuses on individual elements, molecular spectroscopy examines chemical bonds and functional groups, and crystallographic techniques such as X-ray diffraction analyze the arrangement of atoms in crystalline solids.

### 1.2. Molecular Spectroscopy

**Molecular spectroscopy** is a branch of spectroscopy that examines the interaction of electromagnetic radiation with molecules to obtain information about their structural and chemical properties.

When a molecule is exposed to electromagnetic radiation, it may experience a change in its:

- Electronic state, as electrons move to higher energy levels;
- Vibrational state, as atoms oscillate around their equilibrium positions;
- Rotational state, as the molecule rotates around its axis.

Thus, depending on the type of observed transition, the following **Types of Molecular Spectroscopy** are distinguished:
 
1. <u>Rotational Spectroscopy</u>:
 - **rotational transition**: 1 - 20 cm $^{-1}$ - Microwave/Far-Infrared region;
 - Requires a permanent dipole moment (e.g., CO, HCl, HF).
 - Applications: determining molecular bond lengths, isotopic substitution effects.

2. <u>Vibrational Spectroscopy</u>:
 - **vibration transition**: 2000 - 4000 cm $^{-1}$ -  Near Infrared region;
 - Found in the infrared (IR) region and include <a href="https://en.wikipedia.org/wiki/Infrared_spectroscopy" target="_blank">IR scectroscopy</a>  and  <a href="https://en.wikipedia.org/wiki/Raman_spectroscopy" target="_blank"> Raman spectroscopy </a>.
 - Typically involves diatomic or polyatomic molecules with dipole moments.
 - Applications: identifying functional groups in molecules (e.g., OH, NH, CO). *NO two unique molecular structures produce the same infrared spectrum*.

3. <u>Electronic Spectroscopy</u>:
 - **electronic transition** - 10000 - 50000 cm $^{-1}$ - visible/UV region;
 - Applications: provides valuable information about molecular structure, substance identification, material science (investigating optical properties of semiconductors, metals, and insulators), and helps in studying band gaps, defect levels, and carrier concentrations.

 Coupled with different type of transition   (e.g., vibrational and rotational), leading to complex spectra.

### 1.3. Rotation-Vibration Spectrum of diatomic molecules

A simultaneous change in both vibrational and rotational energy levels of a molecule results in a rotational-vibrational spectrum, which can be observed in the infrared (IR) region of the electromagnetic spectrum and typically has a complex structure.

The energy associated with ro-vibrational transitions can be described by combining the energy expressions for both rotational and vibrational motions of a molecule:

- The vibrational energy depends on the bond strength and the masses of the atoms.
- The rotational energy depends on the moment of inertia of the molecule and its rotational constants.

To observe a rotational-vibrational spectrum, a molecule must be  infrared active and typically exist in the gas phase, as this allows the rotational structure to be well-resolved.

The efficiency of photon emission or photon absorption for a free molecule is determined by Einstein coefficients $A_{fi}$ and $B_{fi}$, where $i$ and $f$ refer to the initial and final states of the molecule before and after the photon emission or absoprtion. The coefficient $A_{fi}$ describes the sponteneous emission
$$
A_{fi}=\frac{4}{3}\frac{\omega^3}{\hbar c^3}\vert \vec d_{fi} \vert^2\,.
$$
In the above expression $\omega$ is the angular frequency of emitted light, related to its energy as $E_p = \hbar\omega$. The energy matches the energy difference betweem the initial $E_i$ and final $E_f$ energies of the molecule, $\hbar\omega = E_i - E_f$. Each of the three components of vector of dipole moment $\vec d_{fi}$ is the matrix element of the operator of projection of $\vec d$ on the corresponding axis in space. In calculation, it is more convenient to use the cyclic components, instead of cartesian or spherical components, $d_0$, $d_{+1}$, and $d_{-1}$:
$$
d_0=d_z\\
d_{+1}=-\frac{1}{\sqrt 2}(d_x+i d_y)\\
d_{-1}=\frac{1}{\sqrt 2}(d_x-id_y)\,.
$$
The cyclic components are proportional to spherical harmonics $Y^1_m$ with $m=0,\pm 1$ and can be used to construct $\vert \vec d\vert^2$ in the same way as cartesian components 
$$
\sum_p \vert d_p\vert^2 =\vert \vec d\vert^2\,. 
$$
Therefore, for the evaluation of the Einstein coefficient, one needs to evaluate the three matrix elements 
$
\langle f \vert d_p \vert i \rangle  \,.
$ 

In the derivation, we work within the <a href="https://en.wikipedia.org/wiki/Born%E2%80%93Oppenheimer_approximation" target="_blank"> Born-Oppenheimer approximation</a> and neglect the spin-orbit interaction. For many applications involving light-element molecules, these two approximations are good. In this situation, the wave functions $\vert i\rangle $ and $\vert f\rangle $ of a diatomic molecule are separable in three factors, electronic $\varphi(\mathbf{r_N})$, vibrational $\chi(R)$, and rotational $\cal R(\Omega)$ wave functions. $\mathbf{r_N}$ refers collectively to the coordinates of all $N$ electrons in the molecule, $R$ is the internuclear distance in the molecule, $\Omega$ is the three Euler angles. Specifying all relevant quantum numbers, the initial state is
$$
\vert i \rangle = \varphi_e(\mathbf{r_N})\chi_{eJv}(R){\cal R}^J_{M\Lambda}\,,
$$
where $e$ refers to a particular electronic state of the molecule (in the Born-Oppenheimer approximation), $J$ is the total angular momentum of the molecule, $M$ and $\Lambda$ are its projections on the z axis in the laboratory frame and on the molecular axis, correspondingly. Notice that $\Lambda$ is also equal to the projection of the electronic angular momentum on the molecular axis. The rotational wave function is given in terms of Wigner functions $D^J_{M\Lambda}$
$$
{\cal R}^J_{M\Lambda}(\Omega)=\left[D^J_{M\Lambda}(\Omega)\right]^*\,.
$$
The final state is written in a similar way
$$
\vert f \rangle = \varphi_e'(\mathbf{r_N})\chi_{e'J'v'}(R){\cal R}^{J'}_{M'\Lambda'}\,.
$$

The components of the dipole moment above are given in the laboratory frame (the formula was derived assuming that the photon polarization is in the laboratory frame). Typically, one evaluares the dipole moments in the molecular frame. Therefore, one needs to connect the two frames. It is usually made by transforming $d_p$ in the molecular frame using the techniques of rotational transformation between two frames rotated with respect to each other by the three Euler angles:
$$
d_p=\sum^1_{\pi=-1} \left[D^1_{p\pi}(\Omega)\right]^* d_\pi\,,
$$
where $d_\pi$ are the components of the vector of dipole moment in the molecular frame.
With these defintions, the matrix element $\langle f \vert d_p \vert i \rangle $ is given by
$$
\langle f \vert d_p \vert i \rangle =\frac{\sqrt{(2J+1)(2J'+1)}}{8\pi^2}\sum^1_{\pi=-1} \int d\Omega D^{J'}_{M'\Lambda'}(\Omega) \left[D^1_{p\pi}(\Omega)\right]^* \left[D^J_{M\Lambda}(\Omega)\right]^*   \int dR \chi^*_{e'J'v'}(R) \langle {e'} \vert d_\pi \vert e \rangle \chi_{eJv}(R)\,.
$$
In the above expression, the quantity $\langle {e'} \vert d_\pi \vert e \rangle$ refers to the integral over all electronic coordinates
$$
\langle {e'} \vert d_\pi \vert e \rangle=\int \varphi^*_{e'}(\mathbf{r_N})d_\pi \varphi_{e}(\mathbf{r_N}).
$$
It is usually referred as electronic transition dipole moment, depends on $R$ and is typically evaluated numerically using quantum-chemistry approaches. In the section below, discussing the spectrum of the CO molecule, it is referred as $D(R)$, where the initial $e$ and final states $e'$ are the same ground electronic states.

The integral over Euler angles $\Omega$ is evaluated analytically
$$
 \int d\Omega D^{J'}_{M'\Lambda'}(\Omega) \left[D^1_{p\pi}(\Omega)\right]^* \left[D^J_{M\Lambda}(\Omega)\right]^*=\frac{8\pi^2}{2J'+1}C^{J'M'}_{JM,1d}C^{J'\Lambda'}_{J\Lambda,1\pi}\,,
$$
where $C^{J'M'}_{JM,1d}$ and $C^{J'\Lambda'}_{J\Lambda,1\pi}$ are <a href="https://en.wikipedia.org/wiki/Clebsch%E2%80%93Gordan_coefficients" target="_blank">  Clebsch‚ÄìGordan coefficients</a>. For a given pair of two electronic states $e$ and $e'$ (and $\Lambda$ and $\Lambda'$) only one term in the sum over $\pi$ survives with $\pi=\Lambda'-\Lambda$. Therefore, 
$$
\langle f \vert d_p \vert i \rangle =\sqrt{\frac{(2J+1)}{2J'+1}}C^{J'M'}_{JM,1d}C^{J'\Lambda'}_{J\Lambda,1\Lambda'-\Lambda }  \int dR \chi^*_{e'J'v'}(R) \langle {e'} \vert d_{\Lambda'-\Lambda} \vert e \rangle \chi_{eJv}(R)\,.
$$
and its absoulte value squared
$$
\vert\langle f \vert d_p \vert i \rangle\vert^2 =\frac{(2J+1)}{2J'+1}\left\lvert C^{J'M'}_{JM,1d}\right\rvert^2 \left\lvert C^{J'\Lambda'}_{J\Lambda,1\Lambda'-\Lambda }  \int dR \chi^*_{e'J'v'}(R) \langle {e'} \vert d_{\Lambda'-\Lambda} \vert e \rangle \chi_{eJv}(R)\right\rvert^2\,.
$$

The orientation of the molecule in space is specified by projections $M$ and $M'$. In normal conditions, the orientation is random. It means that one should average over initial orientations $M$. Also, one is usually interested to evaluate the probability of emission/absorption to all possible states of orientations of the molecule, i.e. one needs to evaluate a sum over $M$. Evaluating the sum, one uses the following properties
$$
\sum_{MM'}\left\lvert C^{J'M'}_{JM,1d}\right\rvert^2 = \frac{2J'+1}{3}\,,
$$
so that
$$
\frac{1}{2J+1}\sum_{MM'}\vert\langle f \vert d_p \vert i \rangle\vert^2 =\frac{1}{3} \left\lvert C^{J'\Lambda'}_{J\Lambda,1\Lambda'-\Lambda }  \int dR \chi^*_{e'J'v'}(R) \langle {e'} \vert d_{\Lambda'-\Lambda} \vert e \rangle \chi_{eJv}(R)\right\rvert^2\,.
$$
Taking the sum over all three $p=0,\pm 1$, one obtains the expression for the $A_{fi}2$ averaged over initial orientations (projections $M$) and accounting for all possible final orientations (projections $M'$) of the molecule after the spontaneous emission of a photon with energy $E_p=\hbar\omega$
$$
A_{fi}= \frac{4}{3}\frac{\omega^3}{\hbar c^3} \left\lvert C^{J'\Lambda'}_{J\Lambda,1\Lambda'-\Lambda }\right\rvert^2  \left\lvert\int dR \chi^*_{e'J'v'}(R) \langle {e'} \vert d_{\Lambda'-\Lambda} \vert e \rangle \chi_{eJv}(R)\right\rvert^2\,.
$$

The Einstein coefficient for absorption and induced emission of a photon with an arbitrary polarization and the direction of propagation is given
$$
B_{fi}= \frac{4}{3}\frac{\pi^2}{\hbar^2}\vert \vec d_{fi} \vert^2\,.
$$
Performorming the calculation similar to the above one for the spontaneous emission, one obtains
$$
B_{fi}= \frac{4}{3}\frac{\pi^2}{\hbar^2}\left\lvert C^{J'\Lambda'}_{J\Lambda,1\Lambda'-\Lambda }\right\rvert^2  \left\lvert\int dR \chi^*_{e'J'v'}(R) \langle {e'} \vert d_{\Lambda'-\Lambda} \vert e \rangle \chi_{eJv}(R)\right\rvert^2\,.
$$

### Selection rules

Selection rules define the transition probability from one eigenstate to another eigenstate. 
The Clebsch-Gordan coefficient  $C^{J'\Lambda'}_{J\Lambda,1\Lambda'-\Lambda }$ in the Einstein coefficients is not zero only if $J'=J\pi +1$ and the difference in projections $\vert\Lambda'-\Lambda\vert\le 1$.

## 2. Potential energy curve and wave functions

To compute the **vibrational-rotational energy levels** and **wavefunctions** of a diatomic molecule in a Morse potential with rotational effects, we need to solve the time-independent Schr√∂dinger equation:

$$H \Psi(r)  = E\Psi(r)$$

where the Hamiltonian $H$ is the sum of kinetic energy $T$ and potential energy $V_\mathrm{eff}(r)$.

For describing the anharmonic interaction between the atoms in a diatomic molecule the Morse potential is used:

$$V(r) = D_e \left( 1 - e^{-\alpha(r - r_e)}\right)^2$$

where $D_e$ is the dissociation energy, $\alpha$ determines the width of the potential well, and $r_e$ is the equilibrium bond length.

Based on description the molecule as rigid rotor the rotational energy contribution can be added:

$$
V_\mathrm{eff}(r) = V(r) + \frac{\hbar^2}{2\mu r^2}(J(J+1)-\Lambda^2)
$$

where $J$ is the total angular momentum quantum number of the system, $\Lambda$ is the projection of the electronic angular momentum onto the internuclear axis (for CO,  $\Lambda = 0$), and $\mu$ is the reduced mass of the diatomic system. In a system with two or more interacting particles, the motion can be described in terms of a single effective mass, which simplifies the equations of motion. For a diatomic molecule, the reduced mass is given by

$$\mu = \frac{m_1m_2}{m_1 + m_2}.$$

where $m_1$ and $m_2$ are the masses of the two defferent atoms that form the molecule.

```python
#CO
# in the interface front
Nr = 1000            # total number of grid point
rmin = 0.05          # minimal distance of the grid point
rmax = 20.           # maximal distance of the grid point 

m1 = 12              # mass 1  in u.
m2 = 15.999          # mass 2  in u.

jmax_to_print = 3    # maximum  j to be displayed and computed 
vmax_to_print = 4    #<Nr maximum  j to be displayed and computed 
jmax = 10            #<200, to match experimental data jmax = 30

# in the code
re = 1.128323        #2.13          # in a.u.
De = 90014           # in 1/cm  or 11.16 in eV x 8065.73 1/cm (? 11.26)
W = 0                # in 1/cm
alpha = 2.2992       # in a.u. or 1.8677  in 1/A
vmax = Nr            # maximum to be put in the code


# conversion
mu = m1 * m2/(m1 + m2)    #the reduced mass
mu = mu/au_amu            #*1e-2/6.02e23/9.2e-31
print("mu=",mu * au_amu, "amu = ",mu,"a.u.")

alpha = alpha * au_A 
De = De/Ryd 
re = re/au_A

omega = alpha * np.sqrt(2*De/mu)
xe = omega/(4 * De)
print("omega=",omega*Ryd,"1/cm = ",omega,"a.u.")

r = np.zeros(Nr,float)
delta_r = (rmax - rmin)/Nr
for i in range(Nr):
    r[i] = rmin + delta_r * i

V = np.zeros((Nr,Nr,jmax),float)
psi = np.zeros((Nr,Nr,jmax),float)
E = np.zeros((Nr,jmax),float)
Eplot = np.zeros((Nr,Nr,jmax),float)

for j in range(jmax):
    for i in range(Nr):
        V[i,i,j] = De * (1 - np.exp(-alpha * (r[i] - re)))**2 + W + j * (j + 1)/(2*mu * r[i]**2)

    T = np.zeros((Nr,Nr),float)
    H = np.zeros((Nr,Nr),float)
    P = np.zeros((Nr,Nr),float)

    LBox = delta_r * Nr
    coef = np.pi * np.pi/(mu * LBox * LBox)
    coefdiag = coef * (Nr * Nr + 2.)/6.

    for i in range(Nr):
        T[i,i] = coefdiag
        for ip in range(i + 1,Nr):      
            s = 1.0/(np.sin((ip - i) * np.pi/Nr))**2
            T[i,ip] = (-1)**(i - ip) * coef * s
            T[ip,i] = T[i,ip]

    H = T + V[:,:,j]
    E[:,j], P = la.eigh(H)
    print('Diagonalization done for j=',j)

    for v in range(Nr):
        Eplot[:,v,j] = E[v,j]
        norm = np.dot(P[:,v],P[:,v])
        psi[:,v,j] = P[:,v]/np.sqrt(abs(norm))

    # Setting bound wavefunctions to be real
        maxloc = np.argmax(abs(psi[:,v,j]))
        psi[:,v,j] = psi[:,v,j] * abs(psi[maxloc,v,j])/psi[maxloc,v,j]

    # DVR transformation
        psi[:,v,j] = psi[:,v,j]/np.sqrt(delta_r)
    
print(E)
```

## 3. Observable Quantities

### 3.1. Rotational constants

Rotational constants are fundamental quantities in rotational spectroscopy that characterize the energy levels of rotating molecules. They provide insight into molecular structure, bond lengths, and moments of inertia of molecule.

For a diatomic molecule (for description use the model of the rigid rotor), the equilibrium rotational constant $B_e$  ($v = 0)can be approximated using:

$$
B_e = \frac{\hbar^2}{2 I}
$$

where $ I = \mu r^2$ is the moment of inertia with $r$ - the equlibrium bond length and $\mu$ - the reduced mass of the molecule.

To find the rotational constant $B_v$ averaged over the vibrational state in rovibrational spectroscopy, we evaluate the integral that accounts for the effect of vibrational motion on the molecular rotational constant (the coupling between rotational and vibrational motions in molecules): 

$$B_v=\int dr u_{v,0}^*(r)\left(\frac{\hbar^2}{2\mu r^2}\right)u_{v,0}(r)$$

where, <br>
$u_{v,0}(r)$ is the vibrational wavefunction for the $v$-th vibrational state, describing the probability of finding a molecule at a certain bond length $r$ in the vibrational state $v$. <br> $\frac{\hbar^2}{2\mu r^2}$ represents the rotational energy contribution for the molecule.

```python
Bv = np.zeros(Nr,float)
v_plot = np.zeros(Nr,int)
for v in range(vmax_to_print): 
    v_plot[v] = v
    for i in range(Nr):
        Bv[v] = Bv[v] + delta_r * (1./(2*mu * r[i]**2)) * psi[i,v,0]**2
           
plt.plot(v_plot[0:vmax_to_print],Bv[0:vmax_to_print]*Ryd,'o')
plt.xlabel("$v$",size=15)
plt.ylabel("$B_v$ (cm$^{-1}$)",size=15)
plt.xticks(v_plot[0:vmax_to_print], v_plot[0:vmax_to_print], rotation=0)
```

The rotational energy for a rigid diatomic rotor is quantized:
$$
E_J = B_vJ(J + 1)
$$
where $J$ is rotational quantum number (0, 1, 2, ..), $B_v$ - the rotational constant with a correction factor introduced to account for the distortion of the molecule due to rotation.

### 3.2. Energy spectra

The ro-vibrational bands consist of a series of lines due to rotational transitions ($ \Delta J$) that occur simultaneously with a vibrational transition ($v \rightarrow v'$). According to the selection rules (see more details in the next chapter), allowed rotational transitions are follows:<br>
    - **P-Branch** ($\Delta J = -1$): The molecule loses rotational energy, transitioning as $J \rightarrow J - 1$. <br>
    - **R-Branch** ($\Delta J = +1$): The molecule gains rotational energy, transitioning as $J \rightarrow J + 1$. <br>
    - **Q-Branch** ($\Delta J = 0$):  Not observed in CO, as it is forbidden in diatomic molecules that do not exhibit a dipole moment change.

## 4. Selection rules

<u>Selection rules</u> determine the allowed transitions between different quantum states in a molecule due to absorption or emission of electromagnetic radiation, ensuring that only specific transitions contribute to observed spectral lines. 
 
In quantum mechanics, these rules are governed by the **transition dipole moment**, which defines the probability of a molecule transitioning from one eigenstate to another. The transition dipole moment is considered separately for electronic, vibrational, and rotational transitions.

For rotational-vibrational transitions, the **transition dipole moment**, expressed as:

$$
\mathcal{I}_{v'j',vj}=\int dru^*_{v'j'}(r)D(r)u_{vj}(r)
$$

where:

$u_{vj}(r)$ are the wave functions for a specific quantum state characterized by the vibrational quantum number $v$ and rotational quantum number $j$.<br>
$D(r)$ is the **dipole moment function**, which describes how the molecular dipole moment varies with internuclear distance $r$. This function typically depends on the molecular structure.

To determine the dipole moment $D(r)$ for CO in ground state we use a software package for <i>ab initio</i> quantum chemistry calculations [[MOLPRO](#https://en.wikipedia.org/wiki/MOLPRO)]. 

```python
from scipy.interpolate import interp1d

data = np.loadtxt("./figures_Spectroscopy/CO_dip_atomic_unit.dat")  # download data from file
r_t = data[:, 0]  # bond lengths
D_t = data[:, 1]  # dipole moments

# # Interpolate the D values for for match with r and psi
r1 = r_t[0]
r2 = r_t[-1]

range_interp = np.logical_and(r >= r1, r <= r2)
r_interp = r[range_interp] 

interpolation = interp1d(r_t, D_t, kind='cubic')
D_interpolated = interpolation(r_interp)

plt.plot(r_t, D_t, "." , label = 'Calculated data')
plt.plot(r_interp, D_interpolated, label = 'Interpolated data')

plt.xlabel("$r$, a.u.",size=20)
plt.ylabel("$D(r)$, a.u.",size=20)
plt.legend()
plt.xlim(r1,5.5)
```

Under the Born-Oppenheimer approximation, when rotational contributions are treated separately ($j = j' = 0$), the transition dipole moment simplifies to:
$$
\mathcal{I}_{v'0,v0}=\int dru^*_{v'0}(r)D(r)u_{v0}(r) .
$$

```python
j1 = 0
j2 = 0 

I1 = np.einsum('rv,r,rw->vw',
               psi[range_interp, :vmax_to_print, j1,],
               D_interpolated,
               psi[range_interp, :vmax_to_print, j2]
            )*delta_r
```

The transition dipole moment depends on the overlap of the wavefunctions. A transition is allowed if the transition dipole moment is nonzero, i.e.,
$$\mathcal{I}_{v'j',vj} \neq 0.$$  
This conditions leads to <u> the vibrational selection rule </u>: 
- $\Delta v = \pm 1$ for fundamental transition. 
- $\Delta v = \pm 2, \pm 3, ... $ for overtone transitions.  These transitions are much weaker and occur due to anharmonicity.

```python
import seaborn
s = seaborn.heatmap(I1[0:vmax_to_print,0:vmax_to_print],annot=True)
s.set(xlabel="$(v,j)=(v,0)$", ylabel="$(v',j')=(v',0)$")
```


Selection rules for <u>rotational transitions</u> determine the allowed transitions between rotational states in a molecular system. These transitions are governed by the transition dipole moment integral:

$$ \mathcal{I}_{j'm',jm}=\iint \sin\theta d\theta d\phi \left[\mathcal{Y}_{j'}^{m'}(\theta,\phi)\right]^*\cos\theta\;\mathcal{Y}_j^m(\theta,\phi) .$$

Using the expression, $\cos\theta=2\sqrt{\frac{\pi}{3}}\mathcal{Y}_1^0(\theta,\phi)$ and the property of the spherical harmonics $\left[\mathcal{Y}_j^m(\theta,\phi)\right]^*=(-1)^m\mathcal{Y}_j^{-m}(\theta,\phi)$ , the integral can be rewritten as
$$ \mathcal{I}_{j'm',jm}=(-1)^m(-1)^{m'}2\sqrt{\frac{\pi}{3}}\int_{0}^{\pi}\int_{0}^{2\pi} \sin\theta d\theta d\phi \left[\mathcal{Y}_j^{-m}(\theta,\phi)\right]^*\left[\mathcal{Y}_1^0(\theta,\phi)\right]^*\mathcal{Y}_{j'}^{-m'}(\theta,\phi). $$
This simplifies to

$$\mathcal{I}_{j'm',jm}=2\sqrt{\frac{\pi}{3}}\sqrt{\frac{(2j+1)(2\times 1+1)}{4\pi(2j'+1)}}C_{j010}^{j'0}C_{jm10}^{j'm'} ,$$ 
where, 
$C_{j_1m_1j_2m_2}^{j_3m_3}=\left\langle j_1m_1j_2m_2\right|j_3m_3\left.\right\rangle $ are the Clebsch Gordan coefficients (real coefficients), which determine the coupling of angular momenta;<br> 


From the properties of Clebsch-Gordan coefficients, the integral is nonzero only if: 
- the rotational quantum number satisfies $\Delta J = \pm 1$, and
- the magnetic quantum number satisfies $\Delta m = 0, \pm 1$.  

These selection rules determine the allowed transitions between rotational states in the presence of an electric dipole interaction (see the figure below).

```python
# in development 
from sympy.physics.wigner import clebsch_gordan

I2 = np.zeros((jmax, jmax,2*jmax + 1,2*jmax + 1),float) #here I(j',j,m',m)
I2_plot=np.zeros((jmax,jmax,2*jmax + 1,2*jmax + 1),float) #here I(j',j,m',m)
# I2_plot=1./I2_plot
x_axis_labels = np.zeros(jmax,int)#float)
y_axis_labels = np.zeros(jmax,int)#float)
v = 0
vp = 1
m_plot = 0
mp_plot = 0

for jp in range(jmax):
    y_axis_labels[jp] = jp#round(E[vp,jp]*Ryd)
    for mp in range(-jp,jp + 1):    
        for j in range(jmax):
            x_axis_labels[j] = j#round(E[v,j]*Ryd)

            for m in range(-j,j + 1): 
                #I2[jp,j,imp,im]=clebsch_gordan(j, 1, jp, m_plot, 0, mp_plot)*2*sqrt(pi/3) # l 1 l' m 0 m'
                #print('imp=',imp,jp,mp,'im=',im,j,m)
                C1 = clebsch_gordan( j, 1, jp, 0, 0, 0)#j1,j2,j3,m1,m2,m3
                C2 = clebsch_gordan( j, 1, jp, m, 0, mp)
                A = np.sqrt(3 * (2*j +1)/(4*np.pi*(2*jp + 1)))
                I2[jp,j,mp,m] = A * 2 * np.sqrt(np.pi/3) * C1 * C2
                I2_plot[jp,j,mp,m] = I2[jp,j,mp,m]
```

```python
ifig = 0
mp = 0
m = 0
print("$\mathcal{I}_{j'"+str(mp)+",j"+str(m)+"}$")
for imp in range(-jmax_to_print + 1,jmax_to_print): 
    for im in range(-jmax_to_print + 1,jmax_to_print):
        plt.figure(ifig)
        if(imp == mp and im == m):
            #print(abs(mp),abs(m),jmax_to_print-1)
            s = seaborn.heatmap(I2_plot[0:jmax_to_print,0:jmax_to_print,imp,im],annot=True)
            s.set(xlabel = "$(vj)="+str(v)+"j$", ylabel="$(v'j')="+str(vp)+"j'$")
            s.set(title = "$(m'm)=("+str(imp)+str(im)+")$")
            s.xaxis.tick_top() # x axis on top
            ifig = ifig + 1
```

The <u>selection rules</u> for rotational-vibrational transitions:

 - <u> For Vibrational Transitions</u>: $\Delta v = \pm 1, \pm 2, ...$ The most intense transitions occur for fundamental transitions ($\Delta v = \pm 1$).

 - <u>For Rotational Transitions</u>:  Due to the interaction of the dipole moment with rotational motion, allowed transitions must follow: $\Delta J = \pm 1$.

## 5. IR Rotational-Vibrational Spectrum 

The intensity of a rotational-vibrational transition in molecule between two level for <u>R branch</u> describe by expression:

$$ \mathcal{R}_{1j+1\leftarrow 0j }=\sum_{m=-j}^j(2j+1)e^{-E_{vj}/k_bT}C\left|\left\langle \Psi_{1,j+1,m}\left|\hat{D}(r)\cos\theta\right|\Psi_{0,j,m}\right\rangle\right|^2 ,$$ 

where $\Psi_{0,j,m}$ and $\Psi_{1,j+1,m}$ are the wavefunctions of the initial and final states, respectively, with the quantum numbers $v = 0$, $v = 1$ for vibrational states and $j$, $j + 1$ for rotational states. <br>
The Boltzmann factor $e^{-E_{vj}/k_bT} $ indicates that transitions from lower vibrational states are more likely to occur at lower temperatures, with higher vibrational levels being less populated.<br>
$C=\frac{\pi\nu_{1j+1\leftarrow 0j}}{3\epsilon_0c\hbar}$ is normalise constant. 

```python
kT = 300.  # in Kelvin to be chosen by the user 
v = 0      # to be chosen by the user
vp = 1     # to be chosen by the user

au_K = 3.1577465e5
kT = kT/au_K

Intensity = np.zeros((jmax,jmax),float)
for j in range(jmax):
    for m in range(-j,j + 1):
        for jp in range(jmax):
            Intensity[jp,j] = Intensity[jp,j] + abs(I2[jp,j,m,m] * I1[vp,v])**2
    Intensity[:,j] = Intensity[:,j]/(2 * j + 1)
```

```python
R = np.zeros(jmax,float)
Delta_R = np.zeros(jmax,float)

iplot_R = 0
c = 137.
epsilon0 = 1./(4 * np.pi)
    
for j in range(jmax):
    for jp in range(jmax):
        if(jp - j == 1):
            Delta_R[iplot_R] = E[vp,jp] - E[v,j]
            nu = Delta_R[iplot_R]/(2 * np.pi)
            C = np.pi * nu/(3 * epsilon0 * c)
            R[iplot_R] = Intensity[jp,j] * C * (2 * j + 1) * np.exp(-E[v,j]/kT)
            iplot_R=iplot_R+1
```

The intensity of a rotational-vibrational transition in molecule between two level for <u>P branch</u> describe by expression: 

$$ 
\mathcal{P}_{1j-1\leftarrow 0j }=\sum_{m=-j}^j(2j+1)e^{-E_{vj}/k_bT}C\left|\left\langle \Psi_{1,j-1m}\left|\hat{D}(r)\cos\theta\right|\Psi_{0,j,m}\right\rangle\right|^2 ,
$$
where $\Psi_{0,j,m}$ and $\Psi_{1,j-1,m}$ are the wavefunctions of the initial and final states, respectively, with the quantum numbers $v = 0$, $v = 1$ for vibrational states and $j$, $j - 1$ for rotational states. <br>
$C=\frac{\pi\nu_{1j-1\leftarrow 0j}}{3\epsilon_0c\hbar} $ is normalise constant.

```python
P = np.zeros(jmax,float)
Delta_P = np.zeros(jmax,float)

iplot_P = 0
c = 137.
epsilon0 = 1./(4 * np.pi) 

for j in range(jmax):
    for jp in range(jmax):
        if(j - jp == 1):
            Delta_P[iplot_P] = E[vp,jp] - E[v,j]
            nu = Delta_P[iplot_P]/(2 * np.pi) 
            C = np.pi * nu/(3 * epsilon0 * c)
            P[iplot_P] = Intensity[jp,j] * C * (2 * j + 1) * np.exp(-E[v,j]/kT)
            iplot_P = iplot_P + 1
```

```python
max_P_R = max(max(R[0:iplot_R]),max(P[0:iplot_P]))
plt.plot(Delta_R[0:iplot_R] * Ryd, R[0:iplot_R]/max_P_R,'ob')
plt.plot(Delta_P[0:iplot_P] * Ryd, P[0:iplot_P]/max_P_R,'or')

maxloc_R = np.argmax(R[0:iplot_R]/max_P_R)
maxloc_P = np.argmax(P[0:iplot_P]/max_P_R)
x_R = Delta_R[maxloc_R] * Ryd
y_R = R[maxloc_R]/max_P_R + 0.05 * R[maxloc_R]/max_P_R

x_P = Delta_P[maxloc_P] * Ryd
y_P = P[maxloc_P]/max_P_R + 0.05 * P[maxloc_P]/max_P_R
plt.annotate('$R$', xy = (x_R,y_R),xytext = (x_R,y_R),xycoords='data',size=15)
plt.annotate('$P$', xy = (x_P,y_P),xytext = (x_P,y_P),xycoords='data',size=15)
plt.xlabel("Energy (cm$^{-1}$)",size=20)
plt.ylabel("Relative intensity",size=20)
plt.ylim(0,1.2)
plt.gca().invert_xaxis()
```

The convoluted intensity for each energy point is calculated based on the R-branch and P-branch of a molecular rotational-vibrational spectrum. This step helps to obtain a model that closely matches the experimental data.

```python
# convolution
Ne = 500000
Energy = np.zeros(Ne,float)
Energy = np.linspace(min(Delta_P[0:iplot_P]) + 20/Ryd,max(Delta_R[0:iplot_R]),Ne)
Convoluted_Intensity = np.zeros(Ne,float)
c = 137.
h = 2 * np.pi
mu = m1 + m2
mu = mu/au_amu
A = np.sqrt(mu/(2 * np.pi * kT))
print('T=',kT * au_K,'K ','M=',mu * au_amu,'amu' )
delta_Energy = Energy[1] - Energy[0]
print(Energy[0] * Ryd,'-->',Energy[Ne - 1] * Ryd,'with a step of',delta_Energy * Ryd,' 1/cm')

for ie in range(Ne):
    nu = Energy[ie]/h

    for iplot in range(iplot_R):
        nu0 = Delta_R[iplot]/h
        Convoluted_Intensity[ie] = Convoluted_Intensity[ie] + c/nu0 * A * R[iplot] * np.exp(-mu * (c * (nu/nu0-1))**2/(2*kT)) * delta_Energy
        
    for iplot in range(iplot_P):
        nu0 = Delta_P[iplot]/h
        Convoluted_Intensity[ie] = Convoluted_Intensity[ie] + c/nu0 * A * P[iplot] * np.exp(-mu * (c * (nu/nu0 - 1))**2/(2*kT)) * delta_Energy
```

```python
print(max(Convoluted_Intensity))

plt.plot(Energy * Ryd,Convoluted_Intensity/max(Convoluted_Intensity),'k')
plt.xlabel("Energy, $E_{vj}$ (cm$^{-1}$)",size=20)
plt.ylabel("Relative intensity, $I$",size=20)
for iplot in range(iplot_R):
    y = R[iplot]/max(Convoluted_Intensity) + R[iplot]/max(Convoluted_Intensity) * 0.1
    x = Delta_R[iplot] * Ryd
    y = y/2.
    x = x + x * 0.001
    plt.annotate('$R_{'+str(iplot)+'}$', xy = (x,y),xytext = (x,y),xycoords='data',size=8)
    
for iplot in range(iplot_P):
    y = P[iplot]/max(Convoluted_Intensity) + P[iplot]/max(Convoluted_Intensity) * 0.1
    x = Delta_P[iplot] * Ryd
    y = y/2.
    x = x + x * 0.002
    plt.annotate('$P_{'+str(iplot)+'}$', xy = (x,y),xytext=(x,y),xycoords='data',size=8)
plt.ylim(0,1.2)
plt.gca().invert_xaxis()
```
