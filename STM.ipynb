## 1. Introduction

<strong>Scanning Tunneling Microscopy</strong> (STM) is a type of microscope used for imaging surfaces at the atomic scale. It is based on the principle of <b>quantum tunneling</b>: when a sharp conducting tip is brought extremely close (a few angstroms) to a conducting or semiconducting surface and a voltage is applied, electrons can tunnel through the vacuum gap, generating a tunneling current. This current depends exponentially on the tip-sample distance, enabling STM to resolve individual atoms on a surface [\[1\]](#bibliography).

STM was invented by Heinrich Rohrer and Gerd Binnig in 1981, for which they were awarded the Nobel Prize in Physics in 1986.

STM has widespread applications across physics, chemistry, and materials science, and it has revolutionized our understanding of surface structures, electronic properties, and the intricate world of nanoscale phenomena. As a non-optical technique, STM is an indispensable tool for:

- visualization of individual atoms and molecules on surfaces;
- manipulation of atoms with high precision (e.g., moving single atoms into new arrangements);
- studying surface phenomena such as adsorption, diffusion, and chemical reactions, as well as probing the electronic properties of materials at the atomic level.

These capabilities have opened new frontiers in the exploration of nanomaterials and nanodevices [\[2\]](#bibliography).

STM operates through three primary modes, each designed for different types of surface investigations:

1. **Imaging mode**: Used to obtain topographical images of a surface at atomic resolution. It includes two scanning regimes:

- *Constant-current* mode (Fig. 1(b)): In this mode, a feedback loop maintains a constant tunneling current ($I_t$) between the tip and sample at each lateral position $(x, y)$. Consequently, the tip's $z$-position is continuously adjusted during scanning. The resulting image reflects the topography of the surface at constant current.

- *Constant-height* mode (Fig. 1(c)): In this mode, the tunneling current is measured at a fixed vertical position $z$ of the tip. This mode allows faster scanning but requires a very flat surface to avoid tip-sample contact.

2. **Spectroscopy mode**: In this mode, the resulting tunneling current $I_t$ is recorded as a function of varying parameters such as tip-sample distance $z$ or applied voltage $V$:

- $I(z)$ spectroscopy:  Measures how the tunneling current varies with vertical displacement of the tip $z$ at constant voltage $V$. This is useful for characterizing the quality, sharpness, and cleanliness of the STM tip.

- $I(V)$ spectroscopy: Measures the tunneling current $I_t$ as a function of voltage $V$ at a fixed tip position. This provides crucial insights into the surface electronic structure, including barrier heights, the local density of states (LDOS), molecular motion alysis (for example, identifying vibrational or rotational modes of molecules).

3. **Manipulation mode**: allows for the deliberate manipulation of atoms and molecules on the sample surface using the STM tip. By applying voltage pulses or carefully adjusting the tip-sample distance, it becomes possible to pick up, move, and position atoms or molecules with precision. This capability opens avenues for constructing nanostructures, investigating surface reactions, and studying the behavior of individual atoms and molecules in real time [\[3\]](#bibliography).

## 2. Tunneling a Single Particle Through a One-Dimensional Potential Barrier

The elementary theory of one-electron tunneling through a one-dimensional potential barrier, which forms the basis of STM, is briefly reviewed in this section.  In quantum mechanics, the probability of electron penetration through a potential barrier, $U(z)$ (see Fig. 2(b)), is described by a wave function $\psi(z)$, obtained from the time-independent Schrödinger equation:

\begin {equation}
\frac{-\hbar^2}{2m}\frac{d^2\psi(z)}{dz^2} + U(z)\psi(z) = E\psi(z) \tag{1}
\end{equation}

where $\hbar$ is the reduced Planck’s constant, $m$ is the mass of the electron, $E$ is its energy and $z$ its position.


### 2.1. Rectangular Potential Barrier

To begin, consider a simple rectangular potential barrier through which electrons tunnel from the sample to the tip region.

In the case of a rectangular potential barrier $U(z)$ and E < U  (where $E$ is the particle energy and $U$ is the barrier height), the solution to the Schrödinger equation(1) in each region - the sample (s), barrier (b) and tip (t) - is given by:


1. Region I (sample): $z \leq 0$ and $U(z)= 0:$  
$$\psi_s(z) = e^{ikz} + Ae^{-ikz} \Rightarrow \text{ incoming wave + reflected wave}\tag{2.1}$$

2. Region II (barrier): $0 < z < d $ and $U(z) = U:$ 
$$\psi_b(z) = Be^{-\varkappa z} + Ce^{\varkappa z} \Rightarrow \text{ decaying wave function in the barrier}\tag{2.2}$$

3. Region III (tip): $z \geq d$ and $U(z) = 0:$
$$\psi_t(z) = De^{ikz} \Rightarrow \text{transmitted wave} \tag{2.3}$$

where $ k = \frac{\sqrt{2mE}}{\hbar} $ and $\varkappa = \frac{\sqrt{2m(U - E)}}{\hbar}.$

In each region, the wave function generally consists of waves propagating along the
$z$-axis (incoming, reflected or decaying). In the sample region, we assume the electron approaches the barrier with unit amplitude (the first term), while the second term represents the reflected wave, with the coefficient $A$ characterizing the reflectance from the barrier.  In the tip region, only the transmitted wave is considered, represented by the coefficient $D$.

The coefficients $A, B, C$, and $D$ are determined by applying the wave function continuity conditions at the two interfaces:
$$\psi_s(0) = \psi_b(0)\tag{3} $$
$$\psi_b(d) = \psi_t(d) \tag{4}$$
$$ \frac{d\psi_s(0)}{dz} = \frac{d\psi_b(0)}{dz} \tag{5}$$
$$\frac{d\psi_b(d)}{dz} = \frac{d\psi_t(d)}{dz}\tag{6}

Equations (2)-(6) can be written in the following matrix form:
\begin{align}
    \begin{pmatrix}
        1 & -1 & -1 & 0 \\
        -ik & \varkappa & -\varkappa & 0 \\
        0 & e^{-\varkappa d} & e^{\varkappa d} & -e^{ikd} \\
        0 & -\varkappa e^{-\varkappa d} & \varkappa e^{\varkappa d} & -ike^{ikd} \\
    \end{pmatrix}
    \begin{pmatrix}
        A \\
        B \\
        C \\
        D \\
    \end{pmatrix}
    &=
    \begin{pmatrix}
        -1 \\
        -ik \\
        0 \\
        0 \\
    \end{pmatrix}
    \tag{7}
\end{align}

From Eq.(7) we can find analytical expressions for the coefficients using *Mathematica*:

$$ A = \frac{(k^2 + \varkappa^2) \sinh(\varkappa d)}{2 \cdot 1j \cdot k \cdot \varkappa \cosh(\varkappa d)+(k^2-\varkappa^2) \sinh(\varkappa d)} \tag {8}$$
$$ B = \frac {k(k + 1j \cdot \varkappa) \exp(\varkappa \cdot d)}{2 \cdot 1j \cdot k \cdot \varkappa \cosh(\varkappa d)+(k^2-\varkappa^2) \sinh(\varkappa d)} \tag{9}$$
$$ C = \frac {-k(k - 1j \cdot \varkappa) \exp(-\varkappa \cdot d)}{2 \cdot 1j \cdot k \cdot \varkappa \cosh(\varkappa d)+(k^2-\varkappa^2) \sinh(\varkappa d)} \tag{10}$$
$$ D = \frac{2 \cdot 1j \cdot \varkappa \cdot k \exp(-1j \cdot k * d)}{2 \cdot 1j \cdot k \cdot \varkappa \cosh(\varkappa d)+(k^2-\varkappa^2) \sinh(\varkappa d)} \tag {11} $$

Additionally, Eq. (7) can be solved numerically using Python (e.g., with NumPy), as demonstrated in this notebook.

In the following, we use these expressions to calculate the wave function and plot its <b>real part</b> for a <b>rectangular barrier</b>.

```python
"""

    Determining Constant and Physical Variables:
        m - mass of the electron [kg];
        U_J - potential energy in joules [J];
        U - potential energy in electronvolts [eV];
        E - particle energy [J];
        k - wave vector [1/m];
        kappa - Decay constant [1/m];
        d - width of the potential barrier [m].

        V - applied voltage [V]
        !!! Throughout this notebook, we use SI (the International System of Units)) unless otherwise specified. 
        For reference: 1 eV = 1.6e-19 J


"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad


#Constants
e = 1.6e-19
m = 9.11e-31            
hbar = 1.054e-34        #redused Planck constant in 6.5821e-16 eV*s or 1.054e-34 J*s
h = 2*np.pi*1.054e-34   #Planck constant J*s

# Data input
U = 1    
U_J = U * e 
E = 0.5*U_J

phi_s = 4.7*e # Work function in J sample (Graphite)
phi_t = 5.3*e  # Work function in J tip (platinum)

V = 0.3       
d = 10e-10      

k = (np.sqrt(2 * m * E)/hbar)
kappa = (np.sqrt(2 * m * (U_J - E))/hbar)


print(f"k = {k*1e-9:.4} 1/nm")
print(f"kappa = {kappa*1e-9:.4} 1/nm")
```

```python
"""
    Analytical solutions for A, B, C, D in eq(7) wewe obtained using "Mathematica" 
    and are presented in Eqs. (8)-(11).
"""

F = 2*1j*k*kappa*np.cosh(kappa*d)+(k**2-kappa**2)*np.sinh(kappa*d)

A = (k**2 + kappa**2)*np.sinh(kappa*d)/F
B = (k*(k+1j*kappa)*np.exp(kappa*d))/F
C = (-k*(k-1j*kappa)*np.exp(-kappa*d))/F
D = (2*1j*kappa*k*np.exp(-1j * k * d))/F

print(f"A={A}")
print(f"B={B}, C={C}")
print(f"D={D}")
print(f"R + T = {np.abs(A)**2 + np.abs(D)**2}")
```

```python
"""
    Numerical solutions for A, B, C, D in Eq(7) were obtained using NumPy's.
"""

M = np.array([[1,-1,-1,0], 
              [-1j*k, kappa, -kappa, 0], 
              [0, np.exp(-kappa*d), np.exp(kappa*d), -np.exp(1j*k*d)], 
              [0, -kappa*np.exp(-kappa*d), kappa*np.exp(kappa*d), -1j*k*np.exp(1j*k*d)]])

N = np.array([-1, -1j*k, 0, 0])

x = np.linalg.solve(M,N)

print(f"A= {x[0]}")
print(f"B={x[1]}, C= {x[2]}")
print(f"D={x[3]}")


print(f"R + T = {np.abs(x[0])**2 + np.abs(x[3])**2}")
```

```python
# Define the conditions for energy dependency
def barrier_energy(z, d):
    out = np.zeros_like(z)
    cond = z <= 0
    out[cond] = 0
    cond = np.logical_and(z > 0,  z < d)
    out[cond] = U
    cond = z >= d
    out[cond] = 0
    return out
    
# Define the conditions for wave function dependency
def wave_function(z, d, x):
    out = np.zeros_like(z, dtype=np.complex64)
    cond = z <= 0
    out[cond] = np.exp(1j*k*z[cond]) + x[0]*np.exp(-1j*k*z[cond])
    cond = np.logical_and(z > 0,  z < d)
    out[cond] = x[1]*np.exp(-kappa*z[cond]) + x[2]*np.exp(kappa*z[cond])
    cond = z >= d
    out[cond] = x[3]*np.exp(1j*k*z[cond])
    return out

# Define the range of z values
z_min = -4*d
z_max = (4 + 1)*d

z = np.linspace(z_min, z_max, 1000)  # Adjust the range as needed

# Calculate the energy values for each z
barrier_energy_values = barrier_energy(z, d)
wf_energy_values = wave_function(z, d, x)


# Plot the energy and wave function dependencies
fig, ax1 = plt.subplots()

ax1.set_ylabel('z, $nm$')
ax1.set_xlabel('Energy, eV')
ax1.plot(barrier_energy_values, z*1e9, 'black')

ax1.tick_params(axis='y')
ax1.axvline(x=E/e, color='red', linestyle='--')
ax1.axvline(x=U, color='green', linestyle='--')

# Add labels near the vertical dashed lines
ax1.text(E/e-0.01, z_max*1e9, 'E', color='red', verticalalignment='bottom', horizontalalignment='right')
ax1.text(U-0.01,  z_max*1e9, 'U', color='green', verticalalignment='bottom', horizontalalignment='right')

ax2 = ax1.twiny()  
color = 'tab:blue'
ax2.plot(np.real(wf_energy_values),  z*1e9, color=color)

ax2.set_ylim([-3.9, 4.9])
ax2.set_xlim([-1.45, 1.7])
ax1.set_xlim([-0.005, 1.1])

ax2.set_xticks([])
plt.show()
```

```python
""" 
    checking the continuity of the wave function at the boundaries of a potential barrier

"""

z0 = d
dz = 1e-10 * d
print("At z = 0")
print(1 + x[0])
print(x[1] + x[2])

print("At z = d")
print(x[1]*np.exp(-kappa*d) + x[2]*np.exp(kappa*d))
print(x[3]*np.exp(1j*k*d))

print(f"f at {z0*1e9}nm from left: {wave_function( np.array([z0-dz]), d, x)}, from right: {wave_function( np.array([z0+dz]), d, x)}")
def d_wave_function(z, d, x):
    out = np.zeros_like(z, dtype=np.complex64)
    cond = z <= 0
    out[cond] = 1j*k*np.exp(1j*k*z[cond]) - 1j*k*x[0]*np.exp(-1j*k*z[cond])
    cond = np.logical_and(z > 0,  z < d)
    out[cond] = -kappa*x[1]*np.exp(-kappa*z[cond]) + kappa*x[2]*np.exp(kappa*z[cond])
    cond = z >= d
    out[cond] = 1j*k*x[3]*np.exp(1j*k*z[cond])
    return out
print(f"f' at {z0*1e9}nm from left: {d_wave_function( np.array([z0-dz]), d, x)}, from right: {d_wave_function( np.array([z0+dz]), d, x)}")
```

The transmission coefficient $T$ represents the probability flux of a particle passing through a potential barrier. It is defined as the ratio of the squared amplitudes of the transmitted and incident wave functions. Assuming the amplitude of the incident wave is 1 (as in Eq. (2) for $ z \leq 0$), the barrier transmission coefficient corresponds to the probability flux density of transmitted electrons in the tip region, which is proportional $|\psi_{t}(z)|^2$. 

$$T = \frac{|\psi_{t}(z)|^2}{|\psi_{i}(z)|^2} = DD^* .\tag{12}$$ 

In this case, since the incident wave has unit amplitude, the transmission coefficient simplifies to Eq.(12), where $D$ is the amplitude of the transmitted wavefunction in the final region ($z ≥ d$).


Define the transmission probability as a function of the particle energy $E$ relative to the barrier height $U$.

1. In the <b> Tunneling regime</b>, where the particle energy $E < U$, the wavefunction inside the barrier decays exponentially. The transmission coefficient is given by: 

$$T = \left(\cosh ^2(\varkappa d)+\frac{1}{4} \left(\frac{\varkappa}{k} - \frac{k}{\varkappa}\right)^2 \sinh^2 (\varkappa d)\right)^{-1} \tag{13.1}.  $$

In the limit where $E << U$ and the barrier is wide ($\varkappa d >> 1$), Eq(13.1) simplifies to:
$$T = T_0 \exp\left(-2d\varkappa \right) \tag{13.1}$$
where $T_0 = 4\left[1 + \frac{1}{4} \left(\frac{\varkappa}{k} - \frac{k}{\varkappa}\right)^2\right]^{-1}$, $ k = \frac{\sqrt{2mE}}{\hbar} $ and $\varkappa = \frac{\sqrt{2m(U - E)}}{\hbar}.$

2. Near the <b> Barrier Top</b> ($E = U$), the wavefunction inside the barrier becomes linear, and the transmission coefficient $T$ is given by:

$$T = \frac{4(1 + k^2d^2)}{4 + k^2d^2} \tag{13.2}$$

3. In the <b>Over-barrier regime</b> ($E > U$), the particle has enough energy to propagate over the barrier. The wavefunction remains oscillatory in all regions, and the transmission coefficient becomes:

$$T = \left[1 + \frac{1}{4} \left(\frac{k'}{k} - \frac{k}{k'}\right)^2 \sin^2(k'd)\right]^{-1} \tag{13.3}$$

where $k' = \frac{\sqrt{2m(E - U)}}{\hbar}.$

### 2.2. Non-Rectangular Potential Barrier: Quasiclassical Approximation

To obtain the transmittance of non-rectangular barriers (when the barrier height is slowly changed relatively to the de Broglie wavelenght: $d << \lambda $) one can use the **quasiclassical approximation**, also known as the **WKB (Wentzel–Kramers–Brillouin) approximation** [\[5\]](#Bibliography).

In this approximation, the transmission probability takes the form:

$$ T(E) \approx \exp\left\{ -\frac{2}{\hbar}\int\limits_{z_1 = 0}^{z_2 = d} \sqrt{2m(U(z) - E)dz} \right\}\tag{14} $$

where $E = \frac{1}{2}m{v}^2$.

One can see, that in case of a square barrier, there is no qualitative difference between calculating the transmission coefficient using exact quantum mechanics (Eq. (13)) and using the WKB approximation (Eq. (14)).



```python
""" 
    Dependencies of Transmission from Energy of Particle 
    T_rect - in the case of rectangular barrier (for different regin E < V, E = V, E > V);
    T- in the case of non rectangular barrier (E < V).

    U_J - potential energy in [J];
    E - particle energy [J];
    k - wave vector [1/m];
    kappa - decaying coefficient [1/m];
    d - potential barrier width [m]. 
"""

# for rectangular barrier eq.12-------------------------------------------------------------------------------

E_rect = np.linspace(0.01 * e, U_J + 2*e, 100)
E_rect_reduit = np.linspace(0.01 * e, U_J- 0.1*e, 20)

d = 10e-10
def rectangular(E, n):
    k = np.sqrt(2 * m * E)/hbar

    if E < U_J:
    # E < U
        kappa = np.sqrt(2 * m * (U_J - E))/hbar
        numerator = np.cosh(kappa * d)**2 + 0.25 * (kappa / k - k / kappa)**2 * np.sinh(kappa * d)**2
        T = 1 / numerator
        if n == 1:
            T1 = 4 /(1 + 1/4 *(kappa/k - k/kappa)**2) * np.exp(-2 * d * kappa) #eq.13 for rectangular barrier
            return T1
    elif E > U_J:
    #E > U
        k1 = np.sqrt(2 * m * (E - U_J)) / hbar
        T = 1/(1 + 1/4 * (k1/k - k/k1)**2 * np.sin(d * k1)**2)
    else:
        T = 4*(1 + k**2 * d**2)/ (4 + k**2 * d**2)
    return T

# Transmittance for rectangular potential
T_rect = np.array([rectangular(E, 2) for E in E_rect])
T1 = np.array([rectangular(E, 1) for E in E_rect_reduit]) #eq.13 for rectangular barrier

##########################################################################################################
# Transmittance for non-rectangular barriers eq.14
E = np.linspace(0.01 * e, U_J, 100)

#the type of this function depends on the shape of the barrier
sigma_nonrect = 1e-9
def Potential(z):
    return U_J * np.exp(-z**2 / (2 * sigma_nonrect**2)) if 0 <= z <= d else 0 #Non-rectangular potential function U(z)

# Integrand for the integral
def integrand(z, E):
    potential_energy = Potential(z)
    if potential_energy > E:
        return np.sqrt(2 * m * (potential_energy - E))
    else:
        return 0

# Function to calculate the transmittance T(E) for non-rectangular potential function U(z)
def non_rectangular(E):
    if E >= U_J:
        return 1.0  # For energies greater than the potential barrier, T(E) = 1
    else:
        integral, _ = quad(integrand, 0, d, args=(E)) # type: ignore
        return  np.exp(-2 * integral / hbar)


# Calculate transmittance for non-rectangular potential function U(z)
T =  np.array([non_rectangular(E_) for E_ in E])


plt.plot(E_rect/e, T_rect, color='blue')
plt.plot(E/e, T, color='green')
plt.plot(E_rect_reduit/e, T1, color='blue', marker='o', markersize=4)

plt.axvline(x=U, color='red', linestyle='--')
plt.text(U-0.03, T_rect.max()-0.01, 'U', color='red', verticalalignment='bottom', horizontalalignment='right')


plt.xlabel('Energy, eV')
plt.ylabel('Probability of Transmission, T')
plt.title('Dependence of Transmission on Particle Energy')

plt.grid(True)
plt.show()

```

## 3.  Tunneling Current Density in Metal-Insulator-Metal Systems

To understand the operation of a Scanning Tunneling Microscope (STM) in the Scanning Tunneling Spectroscopy (STS) regime — where the tunneling current as a function of voltage, $I_t(V)$, is measured — it is helpful to examine how the tunneling current in Metal-Insulator-Metal (MIM) systems with arbitrarily shaped potential barriers depends on the applied voltage.

<div style="text-align: center;">
<img src="./figures_STM/Fig_2b_07_03_new.png" alt="Consept ot Tunneling" width="30%"  />

Fig. 2. Model of a MIM system with a non-rectangular potential barrier and an insulating layer of thickness $d$ between the electrodes. The applied bias is denoted by $V$, and $e$  is the elementary charge. The effective barrier width for electrons at the Fermi level, ​$\delta_z$ is defined as $(z_2 - z_1)< d$ . $E_{F}$ represents the Fermi level. $N_1$ and $N_2$ indicate the number of electrons tunneling through the barrier in opposite directions. $\Phi_{t}$ and $\Phi_{s}$ denote the work functions of the tip and sample materials, respectively.
</div>

In a metal–insulator–metal (MIM) system under thermodynamic equilibrium, the Fermi levels in the metals coincide. This equilibrium state implies a balance in electron distribution between the two electrodes. Applying a bias voltage between the electrodes (or the bias energy $eV$) leads to the initiation of a **tunnel current**, namely, the flow of electrons across the insulating barrier.

The tunneling current density $J = e(N_1 - N_2)$ depends on the energy density of states $n(p_z)$ and the transmission probability $T(E_z)$ for electrons tunneling through a potential barrier $U(z)$. To calculate the current density, we need to determine the number of particles passing through the barrier: $N = N_1 - N_2$, where $N_1$ and $N_2$ represent the number of electrons tunneling in opposite direction:

$$ N_1 = \int\limits_0^{E_\infty}{T(E_z)n(p_z)dE_z} \tag{15}$$

where
 $$ n(p_z) = \frac{1}{4\pi^3\hbar^3}\int\limits_{-\infty}^{\infty}\int\limits_{-\infty}^{\infty}{f_1(E)[1 - f_2(E + eV)]dp_xdp_y} \tag{16}$$
and $f_1(E)$, $f_2(E + eV)$ - the Fermi-Dirac functions for the first (sample) and the second (tip) electrodes, respectively.

After integration expression (16) in polar coordinates (where $p_r^2=p_x^2+p_y^2$ and $E = E_z + E_r$) and substituting  the result into (15), we obtain the number of electrons $N_1$ tunneling through the barrier from electrode 1(sample) into electrode 2 (tip)  [\[6\]](#bibliography):
$$N_1 = \frac{m}{2\pi^2\hbar^3}\int\limits_0^\infty {T(E_z)dE_z}\int\limits_0^\infty{f_1(E_z + E_r)[1 -f_2(E_z + E_r + eV)]dE_r} \tag{17}$$

By analogy, the number of electrons $N_2$ tunneling through the barrier from electrode 2(tip) into electrode 1 (sample) [\[6\]](#bibliography):
$$N_2 = \frac{m}{2\pi^2\hbar^3}\int\limits_0^\infty {T(E_z)dE_z}\int\limits_0^\infty{f_2(E_z + E_r + eV)[1 -f_1(E_z + E_r)]dE_r}\tag{18}$$

To simplify the expression for the tunneling current, we introduce the auxiliary functions $\xi_1(E_z)$, $\xi_2(E_z)$, and their difference $\xi(E_z,eV)$, which represent the energy-dependent contributions of electrons tunneling in each direction.

$$\xi_1(E_z) = \frac{me}{2\pi^2\hbar^3}\int\limits_0^\infty{f_1(E)[1 - f_2(E + eV)]dE_r}$$

$$\xi_2(E_z) = \frac{me}{2\pi^2\hbar^3}\int\limits_0^\infty{f_2(E + eV)[1 - f_1(E)]dE_r}$$

$$\xi(E_z,eV)=\xi_1-\xi_2 = \frac{me}{2\pi^2\hbar^3}\int\limits_0^\infty{\left[f_1(E) - f_2(E + eV)\right]dE_r}

Then the tunneling current density in an MIM system for a potential barrier with an arbitrary shape shaped potential barrier takes the form:

$$ J = \int\limits_0^{E_\infty}{T(E_z)\xi(E_z,eV)dE_z} \tag{19}

According Fig.2, the potential barrier $U(z)$ can be be expressed as $U(z) = \mu + \varphi(z)$. Then, by integrating equation (14) using the approximation $\int_{z_1}^{z_2}{\sqrt{f(z)dz} = \beta\sqrt{\bar f \delta_z}}$ form [\[6\]](#bibliography) we obtain:

$$T(E_z) \propto \exp{\left\{-A \delta_z \sqrt{\mu + \bar \phi(z) - E_z}\right\}} \tag{20} .$$

Here, $\bar \phi$ is the average barrier height relative to the Fermi level of the negative electrode. In the case of vacuum barrier, it is determined by the work functions of the two electrodes. $\mu$ is chemical potential, which coincides with the Fermi level in metals at temperatures close to absolute zero. The factor $A$ is definded in Eq.(23). 

At  $T = 0 K$ the function $\xi(E_z)$ takes the form:

$$\xi(E_z) = \frac{me}{2\pi^2\hbar^3}\begin{cases}
                                        eV, & E_z \in \left[0; \mu - eV \right];\\
                                        \mu - E_z, & E_z \in \left[\mu - eV; \mu \right];\\
                                        0, & E_z > \mu
                                    \end {cases} \tag{21} $$



By substituting equations (20) and (21) into equation (19) and performing the integration, we obtain an approximate expression for the **tunneling current density as a function of the applied voltage** in a metal–insulator–metal (MIM) system with an arbitrarily shaped barrier (for more details, refer to [\[6\]](#bibliography),[\[7\]](#bibliography))

$$ \boxed {J = \frac{J_0}{\delta_z^2}\left\{\bar \phi \exp\left[{-A\delta_z\sqrt{\bar \phi}}\right] - (\bar \phi + eV) \exp\left[-A\delta_z\sqrt{\bar \phi + eV}\right] \right\}}\tag{22}$$

where   
        $$ J_0 = \frac{e}{4\pi^2\beta^2\hbar}, \quad{    }
         A = 2\beta\sqrt{\frac{2m}{\hbar^2}}\tag{23}$$ 

Here,
 - $\beta$ is a correction factor that depends on the applied bias (see [\[7\]](#bibliography)),  
 - $\bar\phi$ is the average barrier height, defined as $\bar\phi = \frac{1}{\delta_z}\int_{z_1}^{z_2}{\phi(z)\,dz}$,
 - $\delta_z$ is the width of the potential barrier for electrons at the Fermi level,  
 - $V$ is the voltage applied between the electrodes.

**Equation (22)  is valid for any potential barrier shape that satisfies the WKB approximation, provided the average barrier height $\bar{\phi}$ is known.  It represents the net tunneling current density as the difference between two opposing current components: $J_0\bar \phi \exp({-A\delta_z\sqrt{\bar \phi}})$, corresponds to electrons tunneling from electrode 1 to electrode 2, and $ J_0(\bar \phi + eV) \exp(-A\delta_z\sqrt{\bar \phi + eV})$, represents tunneling in the reverse direction, from electrode 2 to electrode 1. (see, Fig.2).**

Obtain analitical expressions for tunneling current $I(V)$ in differenr intervals of applied voltage (Fig.3):

### 3.1. Low-Voltage Range

At low voltages $\bar \phi >> eV$, expression (22) can be simplified [\[6\]](#bibliography): 

$$ J_L = \frac{J_0}{\delta_z} eV\left[\frac{A \delta_z \sqrt{\bar \phi}}{2} - 1 \right] \exp \left[ -A\delta_z \sqrt{\bar \phi}\right] \tag{24}$$

Since $\frac{A \delta_z \sqrt{\bar \phi}}{2}  >> 1$, eq.(24) reduces to:

$$ \boxed {J_L = \frac{J_{0L}}{\delta_z} \sqrt{\bar \phi}V \exp \left[ -A\delta_z\sqrt{\bar \phi}\right] \tag{25} }$$

where $$ J_{0L} = \frac{e^2\sqrt{2m}}{4\pi^2\hbar^2\beta} \tag{26} $$

The energy diagram for this case is shown on Fig.3(a), where we can see that $\delta_z = d$ and $\bar \phi = (\Phi_s + \Phi_t) / 2$. The coefficient $\beta$ was defined in [\[7\]](#bibliography), where it was noted that for $eV \simeq 0$, $\beta$ takes the value of unity. The coefficient $A$ is defined in eq.(23).

Thus, in this case, equation (25) shows that $J$ is a linear function of $V$ (see section 1 on the Fig.4.).

### 3.2. Intermediate-Voltage Range

If $eV < \Phi_t$ we can see from Fig.3(b) that $\delta_z = d$ and $\bar \phi = (\Phi_s + \Phi_t - eV)/2$. In [\[7\]](#Bibliography) it is shown, that for this case the tunneling current-voltage relation is given by

$$ \boxed {J_I = \frac{J_{0I}}{\delta_z^2} \left\{\left(\Phi_s - \frac{eV}{2}\right) \exp{\left(-A d \sqrt{\Phi_s -  \frac{eV}{2}}\right)} -
              \left(\Phi_s + \frac{eV}{2}\right) \exp{\left(-A d \sqrt{\Phi_s + \frac{eV}{2}}\right)} \right\}}, \tag{27}$$

where
$$ J_{0I} = \frac{e}{4\pi^2\beta^2 \hbar},$$

The value of $A$ is defined above in eq.(23) and $ \beta \approx 1$ [\[6\]](#bibliography). Using eq.(27) the current-voltage characteristic was plotted in Fig.4 (in red). 

#### 3.3. High-Voltage Range

Figure 3(c) illustrates the energy diagram for the case, when $eV > \Phi_t$. From this figure let's define 

$ \delta_z = d  \Phi_s(\Phi_s - \Phi_t + eV)$  and

$\bar \phi = \Phi_s/2$.

Substituting this values in eq.(18), we obtain expression for the tunneling current density in high voltage range case:

$$ \boxed {J_H = \frac{J_{0H}}{\Phi_s} \left\{ \exp{\left[ -A_H\Phi_s^{3/2}\right]} - \left( 1 + \frac{2eV}{\Phi_s}\right) \exp{\left[ -A_H\Phi_s^{3/2} \sqrt{1 + \frac{2eV}{\Phi_s}} \right]} \right\}}, \tag{28}  $$

where $ F = V/d $ defines the electric field strength in the insulator, the coefficient $J_H$  and $A_H$ are described by the following expressions:

$$ J_{0H} = \frac{e^3F^2}{8\pi^2\hbar\beta^2}$$

$$ A_H = \frac{ 2 \beta \sqrt{m}}{eF\hbar}$$

In this case, it is necessary to determine a value of the correction factor $\beta$, as described in [\[6\]](#Bibliography) 
$$\beta = 1 - \left[ \frac{eV/d^2}{\delta_z}\right]\int\limits_0^{\delta_z = d\Phi_s/eV} (\delta_z/2 - x)^2dx/(\Phi_s/2)^2 = 1 - 1/24 = 23/24 $$



Using equations (25)-(28) plot current-voltage characteristic (Fig.4) for case of carbone electrode 1(sample) with $\Phi = 4,7 eV$ and platinum electrode 2 (tip) with $\Phi = 5,3 eV$ at $d = 5 \AA$ and contact area $ S = 10^{-17} m^2 $. 

! To find the tunnel current in [$A$] you need the tunnel current density [$A/m^2$] to multiply by the contact area [$m^2$].

```python
"""

    Plot the current-voltage dependence for carbon-insulator-platinum barrier structure 
    with width of barrier d = 5e-10 m and contact area S.

    According to [6] and [7]: d, phi and beta depend on the applied voltage.

    Use the basic formula (22) 



    Using equations (21)-(24) plot current-voltage characteristic for case of 
        phi_s = 4,7 * 1.6e-19 J - work function of carbone - electrode 1(sample);
        phi_t = 5,3 1.6e-19 J - work function of platinum electrode 2 (tip) with ;
        d = 5 e-10 m - distance between electrodes 
        S = 10-17 m^2 - contact area S = 10-17 m^2 - a typical benchmark for a good STM tip is a radius of curvature of 50 nm or less at the tip apex. 
        V -  applied voltage in V
"""
d = 1e-9
S = 1e-18

V = np.linspace(0.01, 7, 100)

delta = np.zeros_like(V)
phi = np.zeros_like(V)
beta = np.zeros_like(V)

# Low-Voltage Range : phi >> eV
cond1 = V < 1
delta[cond1] = d 
phi[cond1] = (phi_s + phi_t) / 2
beta[cond1]= 1

# Intermediate-Voltage Range : eV < phi
cond2 = V <= phi_t / e
delta[cond2] = d 
phi[cond2] = (phi_s + phi_t - e*V[cond2]) / 2
beta[cond2] =  1 

# High-Voltage Range: eV > phi_t
cond3 = V > phi_t / e
delta[cond3] = d * phi_s / (phi_s - phi_t + e*V[cond3])
phi[cond3] = phi_s / 2
beta[cond3] = 23/24


# Low-Voltage Range : phi >> eV
A1 = (2 * beta * np.sqrt(2*m)) / hbar #eq.19
J_L1 = ((e**2) * np.sqrt(2*m)) / (beta * (2*np.pi*hbar)**2) #eq.22
J_1 = J_L1  * (V * np.sqrt(phi) / delta) * np.exp(-A1 * delta *np.sqrt(phi))  #eq.21
J_1[~cond1] = np.nan


# Intermediate-Voltage Range : eV < phi
A2 = (2 * beta * np.sqrt(2*m)) / hbar #eq.19
J_L2 = e / (2*np.pi*h) #eq.
J_2 = (J_L2/delta**2) * ((phi_s - e*V/2) * np.exp(-A2 * delta * np.sqrt(phi_s - e*V/2)) -
              (phi_s + e*V/2) * np.exp(-A2 * delta * np.sqrt(phi_s + e*V/2))) #eq.23
J_2[V < 0.5] = np.nan
J_2[cond3] = np.nan


# High-Voltage Range: eV > phi_t

F = V/delta
A3 = 2 * beta * phi_s**(3/2) * np.sqrt(m)/(e*F*hbar)
J_L3 = e**3 * (F/beta)**2 / (8*np.pi*h*phi_s)
J_3 = J_L3 * (np.exp(-A3)
    - (1 + (2*e*V)/phi_s)*np.exp(-A3 * np.sqrt(1 + (2 * e * V) / phi_s))) #eq.24

J_3[cond1] = np.nan
J_3[cond2] = np.nan


# Plot each interval's I(V) dependency

plt.scatter(V, J_1 * S, label="Low-Voltage Range", color='blue')
plt.scatter(V, J_2 * S, label="Intermediate-Voltage Range", color='red')
plt.scatter(V, J_3 * S, label="High-Voltage Range", color='green')

# Add labels and legend
plt.xlabel('Voltage (V)')
plt.ylabel('Current (I)')
plt.title('Current-Voltage Dependency')
plt.legend()
# plt.ylim([0,2e-6])

# Show the plot
plt.grid(True)
plt.show()
```

## 4. Modelling of the Scanning Process

### 4.1. Modeling STM Tip Scanning Trajectory in Constant-Current Mode

```python
"""
        Analysis of the tunneling current as a function of the distance between the tip and the sample surface (Fig.1c);
        The expression [25] for the tunneling current through a one-dimensional potential barrier is used.

        To define a surface profile geometry modify the surface_profile(x) function.
"""
V = 0.3 # Applied voltage in eV
beta = 1
J0 = 7.879038e+13
A = (2 * beta * np.sqrt(2 * m)) / hbar
phi = (phi_s+phi_t)/2
# Surface profile function sinc^2(x)

# Parameters for scaling and shifting
width_scaling_factor = 5e-9 # m Adjust this value to widen or narrow the peak
height_scaling_factor = 0.3e-10  # m Adjust this value to lower or raise the entire function

##############################################################################
def surface_profile(x):
    return height_scaling_factor * np.sinc(x / width_scaling_factor)**2  # sine wave profile

##############################################################################

# Generate x values for the surface profile
x_values = np.linspace(-10e-9, 10e-9, 100)

profile = surface_profile(x_values)
z0 = np.max(profile) + 1e-10 # distance between the tip and sample's surface in this mode is constant = 1e-10
z = z0 - profile 


# Calculate tunneling current at each position x
tunneling_current_surface = (J0 / z) * np.sqrt(phi)* V * np.exp( - A * z *np.sqrt(phi)) 
print (f"tunnel curren density J = {V}")


# Create a figure with 3 subplots stacked vertically
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, sharex=True)

# Plot: The surface's cross section and tip's thrajectory:  I-const mode of STM (Fig.1(b))
ax0.plot(x_values/1e-10, profile/1e-10, color='black') #label='Distance between tip and surface',
ax0.plot(x_values/1e-10, z0/1e-10 * np.ones_like(x_values), label='Trajectory of tip', color='green' )  # Plotting z0(x)
ax0.set_ylabel('z, $\AA$')
ax0.set_yticklabels([])
ax0.legend(loc='upper right', bbox_to_anchor=(1, 1))

# Plot: Distance between tip and surface
ax1.plot(x_values/1e-10, z/1e-10, label='Distance between tip and surface', color='blue') #
ax1.set_ylabel('$\delta$ z, $\AA$')
ax1.set_yticklabels([])
ax1.legend(loc='upper right', bbox_to_anchor=(1, 1))

# Plot the results for power-law relationship on the second subplot
ax2.plot(x_values/1e-10, tunneling_current_surface *S * 1e-9, label='Tunneling Current', color='red')
ax2.set_xlabel('x, $\AA$ ')
ax2.set_yticklabels([])
ax2.set_ylabel('$I_{t}$, nA')
ax2.legend(loc='upper right', bbox_to_anchor=(1, 1))

plt.show()
```

### 4.2 Analysis of Tip-Sample Distance Influence on Tunneling Current

In *constant-height* (Fig.1(c)) or *constant-current* (Fig.1(b)) scanning modes, changes in the vertical position $z$ of the tip affect the tunneling current throughout the scanning process. Any variation in the surface topography alters the tip – sample distance, which, in turn, leads to corresponding changes in the tunneling current.

The typical bias voltage range in STM is from -2 eV to +2 eV. Considering that the materials commonly used for the tip and sample have work functions in the range of 4 – 6 eV, this satisfies the condition 
$ \Phi >> eV, allowing us to use expression (21) for modelling:

$$ J_L = \frac{J_{0L}}{d} \sqrt{\bar \phi}V \exp \left[ -\frac{2\sqrt{2m}}{\hbar}d\sqrt{\bar \phi}\right] $$

$J_{0L}$  is a prefactor determined by equation (22);

$d$  is the tip-sample distance;

$m_{e}$  is the effective mass of the tunneling electrons;

$\bar \phi$  is the work function, in this case, defined as the average value of the work functions of the sample and the tip, respectively: $(\phi_s + \phi_t) / 2$;

$V$  is the applied voltage;

$\hbar$  is the reduced Planck constant

```python
"""
   Dependence of the tunneling current on barrier width.
    
"""
e = 1.6e-19
beta = 1
S = 1e-17

J0 = ((e**2) * np.sqrt(2*m)) / (beta * (2*np.pi*hbar)**2) # Eq.(26)

# phi = 4.5 *e  # Work function eV
phi = (phi_s + phi_t)/2

V = 1 #Applied voltage in V

k = np.sqrt(2 * m * (phi - e*V))/hbar # is the wave vector in 1/A, if V is low. 

A = (2 *beta* np.sqrt(2*m)) / hbar #Eq.(23)

# Calculate the varying tip-sample separation (distance) from the surface profile
d = np.linspace(5e-10, 10e-10, 100)

# Eq.(25)
tunneling_current = (J0 / d) * np.sqrt(phi/2)* V * np.exp( - A * d *np.sqrt(phi/2)) # redused Planck constant in 6.5821e-16 eV*s or 1.054e-34 J*s

print(f"k = {k:.4} 1/m, A={A:.4}, J0={J0:.7}")

# Plot the results
plt.plot(d*1e10, tunneling_current * S, label='Tunneling Current')
plt.xlabel('d, $\AA$')
plt.ylabel('$I_{t}$, A')
plt.title('Tunneling Current vs Barrier width')
plt.legend()
plt.grid(True)
plt.show()
```

As seen, the tunneling current decays exponentially with the gap distance $d$, highlighting the high sensitivity of the STM method. For a typical work function (~ 4.5 eV), the tunneling current decreases by a factor ~10 for every 0.1 nm increase in $d$, implying a change by a factor of **~1000** over a typical atomic diameter (~0.3 nm). T

To estimate the **sensitivity of STM to vertical variations**, one typically consider the method's ability to detect small changes in the tunneling current $ \Delta I$ resulting from slight variations in the tip–surface distance $\Delta d$.
 

$$ Sensitivity = \frac{\Delta I}{\Delta d}$$

```python
"""
    Determining the vertical displacement sensitivity of the method    
"""

phi = 4.5 * e
d = 5e-10
delta_d = 3e-20

J0 = 7.87903797e+13 #value taken from previousncalculation
V = 0.3             # Applied voltage in V

tunneling_current = (J0 / d) * np.sqrt(phi/2)* V * np.exp( - A * d *np.sqrt(phi/2))

tunneling_current_2 = (J0 / (d+delta_d)) * np.sqrt(phi/2)* V * np.exp( - A * (d+delta_d) *np.sqrt(phi/2))


print(f'tunneling_current = {tunneling_current *S} A')
print(f'tunneling_current_2 = {tunneling_current_2*S}A') 
print(f'Sensitivity = {(tunneling_current - tunneling_current_2)*S/delta_d} A/m')
```

## 5. Some Aspects of STM Application

The rapid advancement of materials science, nanotechnology, and biotechnology requires ongoing development in high-resolution instrumentation, fabrication methods, and analytical tools for manipulating and characterizing matter at atomic and molecular scales. The Scanning Tunneling Microscope (STM) is one such tool that addresses these needs.

STM is a non-optical, surface-sensitive instrument that operates based on the principle of electron quantum tunneling. It provides atomic-scale resolution (~2 Å) in three dimensions and is used to analyze surface topography, local density of states, atomic-scale defects, grain boundaries, and variations in chemical composition.

A distinctive capability of STM is its ability to manipulate individual atoms and molecules, allowing for control at the atomic level. However, its reliance on the electrical conductivity of both the sample and the tip limits its use to conductive or semiconductive materials.


The following are examples of STM applications in scientific research and technological development:

1) Application of STM in Electrocatalysis and Electrochemistry: STM enables atomic-level characterization of electrocatalyst surfaces and the study of electrochemical interfaces with reactants, even under ultra-high vacuum or harsh conditions. This capability is demonstrated in the work by [\[12\]](#bibliography). 

2) The use of Scanning Tunneling Microscopy (STM) plays a key role in the development of technologies across various scientific and technological fields. Processes such as nanoparticle synthesis, biological component deposition, ultra-thin film growth, and surface engineering require careful surface preparation, precise control of surface morphology, and rigorous evaluation of results at the atomic scale.
For example, in [\[13\]](#bibliography), a method for preparing the Cu(111) surface for catalysis and the growth of two-dimensional materials was considered. In this study, the morphological and structural changes of the Cu(111) surface after treatment by ion bombardment and annealing in the temperature range of 300–740 °C were systematically investigated using in situ low-temperature scanning tunneling microscopy. 

3) The first quantum calculations using individual titanium atoms on a surface were successfully performed by controlling them with microwave signals emitted from a scanning tunneling microscope (STM) [\[14\]](#bibliography). This breakthrough technology holds significant potential for the development of a new type of quantum computer. The researchers managed to control the spin of a single electron in one of the titanium atoms by directing precisely tuned microwave signals from the STM tip [\[15\]](#bibliography). By adjusting the microwave frequencies, they were also able to induce interactions between the spin of that atom and the spins of two neighboring titanium atoms. Through this precise manipulation of atomic spins, the team implemented a simple two-qubit quantum operation and successfully read out the results.

4) In [\[16\]](#bibliography) discusses the use of scanning probe microscopy, including STM, for characterizing biomolecules such as DNA and Individual Biopolymers by resolving the surface morphology and electronic properties of DNA strands deposited on conductive substrates. 

<b>Comparing STM with Other Microscopy Techniques:</b>

| Technique                             | Resolution              | Imaging Principle                                        | Sample Requirement                  | Information Provided              | 
|---------------------------------------|-------------------------|----------------------------------------------------------|-------------------------------------|-----------------------------------|
| STM (Scanning Tunneling Microscope)   | Atomic-scale (~0.1 nm)  |Quantum tunneling current measurement                     |Conductive or semiconductive surfaces| Topography, electronic structure  |
| AFM  (Atomic Force Microscopy)        | Atomic-scale (~0.1 nm)  |Measures forces between tip and sample surface            |Conductive or insulating             | Topography, force interactions    |
| SEM (Scanning Electron Microscopy)    | ~1–10 nm                |Electron beam scans surface, detects secondary electrons  |Conductive or coated samples         | Surface morphology                |
| TEM (Transmission Electron Microscopy)| Sub-atomic (< 0.1 nm)   |Electron beam transmitted through thin sample             |Ultra-thin samples                   | Internal structure                |

Recent progress in Scanning Tunneling Microscopy has significantly broadened its capabilities through improved resolution, environmental adaptability, and atomic-level precision. Advances such as cryogenic, ultrafast STM (combined with terahertz electromagnetic pulses), or integration with complementary methods (like AFM and spectroscopy), have extended its applications beyond surface science into emerging areas like catalysis, molecular electronics, and biomolecular analysis [\[17\]](#bibliography).
