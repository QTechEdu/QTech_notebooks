## 1. Photoionization cross section

We cosider the process of ionization of an atom. The origin of energy is chosen to be at energy of ionization $E_{th}$: $E_{th}=0$. Therefore, the initial bound state $\vert i\rangle$  of the atom has a negative energy $E_i<E_{th}$, while after ionization, in the final state $\vert f\rangle$ , when one of the electrons is at an infitite distance fromt the atom,  the energy of the system is positive $E_f>E_{th}$.

For the simplicity of the discussion, it is assumed that only one electron in the outer electronic shell participates in the process. The assumption gives the exact result for one-electron systems (such as H or He $^+$ ). The assumption is a very good appoximation for alkali atoms and for many situations of excited states of other atoms. However, the approximation is not accurate for such atoms as He or Ne in their ground electronic state. 

It is also assumed that the $z$-axis of the laboratory reference frame is chosen along the axis of polarization of the incident light. In this situation, the photoionization (PI) cross section is given by the formula
$$\sigma^{PI}=\frac{4\pi^2 e^2}{c}\omega \left| \hat z_{fi} \right|^2\ ,$$
where $\hat z$ is the $z$-coordinate of the active electron. In the general case, a sum over the $z$ components of all electrons should be evaluated.

In the above expression $\vert i\rangle$ and $\vert f\rangle$ are initial and final states of the system. Because one-electron approximation is used, the wave functions of  $\vert i\rangle$ and $\vert f\rangle$ are written as

$$\vert i\rangle=R_i(r) Y_{l_i m_i}(\theta,\phi) ,$$
$$\vert f\rangle= R_f(r) Y_{l_f m_f}(\theta,\phi)$$
and the $z$ component of the electron is represented by spherical coordinates
$$\hat z = r\cos(\theta)=\sqrt{\frac{4\pi}{3}}r Y_{1,0}(\theta,\phi) ,$$


$$
\hat z_{fi}=\langle R_f \vert r \vert R_i \rangle \sqrt{\frac{4\pi}{3}}\langle  Y_{l_f m_f} \vert Y_{10} \vert  Y_{l_i m_i} \rangle  .
$$
The angular part of the integral in the matrix element $\hat z_{fi}$ is evaluated analytically
$$
\langle  Y_{l_f m_f} \vert Y_{10} \vert  Y_{l_i m_i} \rangle =\sqrt{\frac{2l_i+1}{2l_f +1}} \sqrt{\frac{3}{4\pi}} C_{l_i 0 1 0}^{l_f 0}C_{l_i m_i 1 0}^{l_f m_f}\,.
$$
Using the notation $r_{fi}=\langle R_f \vert r \vert R_i \rangle$, $\hat z_{fi}$ becomes 

$$
\hat z_{fi}=r_{fi}\sqrt{\frac{2l_i+1}{2l_f+1}}C_{l_i 0 1 0}^{l_f 0}C_{l_i m_i 1 0}^{l_f m_i}\delta_{m_i m_f}\,.
$$

The above derivation is made for a given initial state $\vert i\rangle $ of the atom and a given final state of  $\vert f\rangle$ of the ion, and for a polarization of the EM-field along $z$-axis in laboratory frame.

If the initial polarization is random and one is interested in the PI cross section into all possible polarizations of the final state, one should average the PI cross section over initial states and make a sum over final states, i.e.

$$
    \frac{1}{2l_i+1}\sum_{m_i,m_f}\vert z_{fi} \vert^2=\frac{1}{3}\vert C_{l_i~0;1~0}^{l_f~0}\vert^2 \vert r_{fi}\vert^2 ,
$$,

so that such a PI cross section averaged over polarizations is

$$
\sigma^{PI}=\frac{4\pi^2 e^2}{3c}\omega \vert C_{l_i~0;1~0}^{l_f~0}\vert^2 \vert r_{fi}\vert^2\ ,
$$

without averaging over initial polarizations but summed over final polarizations is
$$
\sigma^{PI}_{m_i}=\frac{4\pi^2 e^2}{c}\frac{2l_i+1}{2l_f +1}\omega \vert C_{l_i~m_i;1~0}^{l_f~m_i}\vert^2 \vert C_{l~0;1~0}^{l'~0}\vert^2 \vert r_{fi}\vert^2\ .
$$


Radial functions $R_i=\frac{\chi_i}{r}$ and $R_f=\frac{\chi_f}{r}$ are obtained solving the radial Schrodinger equation
$$
-\frac{\hbar^2}{2m_e}\frac{d^2\chi}{dr^2}+\left(\frac{\hbar^2 l(l+1)}{2m_er^2} +V(r)\right)\chi=E\chi\
$$
for the corresponding ($E_i<0$ and $E_f>0$) energies. The potential of interaction is a pure Coulomb potential $V(r)=-\frac{Ze^2}{r}$ if the atom has only one electron, such as H, He $^+$, etc. For multi-electron atoms,  $V(r)$ behaves as $-\frac{Ze^2}{r}$  only for large distances $r$.

At large values of $r$, the bound-state wave function $\chi_i(r)\to 0$, while  $\chi_f(r)$ oscillates. For pure Coulomn field, the asymptotic behavior of $\chi_f(r)$  is 
$$\chi_f(r)\to \sqrt{\frac{2}{\pi k}}\sin\left(kr+\frac{\ln (r)}{k}+\sigma_C\right) .$$


## 2. Coulomb functions for negative enegies

<h3>Whittaker functions</h3>

Two Whittaker functions **whitm()** and **whitw()** are solutions of the equation

$$\frac{d^2w}{dz^2} + \left(-\frac{1}{4}+\frac{k}{z}+ \frac{(\frac{1}{4}-m_w^2)}{z^2}\right) w = 0 $$


***

To solve the Schrodinger equation for the Coulomb potential
$$
-\frac{\hbar^2}{2m_e}\frac{d^2\chi}{dr^2}+\left(-E+\frac{\hbar^2 l(l+1)}{2m_er^2}-\frac{Ze^2}{r}\right)\chi=0
$$
with $E<0$ we transform it into a standard form where known mathematical functions, such as the Whittaker functions (see above) can be applied.   

After the change $r=az$ and $\chi=bw$ the Schrodinger equation is transformed to
$$
\frac{\hbar^2 b}{2m_ea^2}\frac{d^2w}{dz^2}+\left(bE-\frac{\hbar^2 b l(l+1)}{2m_ea^2r^2}+\frac{Ze^2b}{ar}\right)w=0
$$
with $bE=-1/4$ ($b>0$) and $\frac{\hbar^2 b}{2m_ea^2}=1$ it is brought to the top one


$$\frac{d^2w}{dz^2} + \left(-\frac{1}{4}+\frac{Ze^2}{\hbar}\sqrt{\frac{m_e}{-2E}}\frac{1}{z}- \frac{l(l+1)}{z^2}\right) w = 0$$
i.e.
$
k=\frac{Ze^2}{\hbar}\sqrt{m_e /(-2E)}$ - effective quantum number 

and 

$
m_w=\sqrt{l(l+1)+1/4}.
$

The scaling factor $a$ for coordinate is
$
a=\frac{\hbar}{2}\sqrt{1 /(-2m_eE)}.
$

Below, Whittaker functions for E=energy are calculated and plotted. The norm is calculated.

```python
from mpmath import mp, whitm, whitw
import matplotlib.pyplot as plt
import numpy as np


mp.dps = 15; #precision 25 digits
mp.pretty = True
energy_i = -1./5
l_partial_i = 0.

k = np.sqrt(1./(-2 * energy_i)) #effective quantum number
mw = np.sqrt(l_partial_i * (l_partial_i + 1) + 0.25)
scale = np.sqrt(1./(-8. * energy_i)) #the scaling factor a for coordinate

npoints = 1000
r_coor = np.linspace(0.01,10., npoints)
y1 = np.zeros( npoints)
y2 = np.zeros( npoints)
norm = 0.

for i in range(npoints):
  z = r_coor[i]/scale
  y1[i] = whitm(k, mw, z) # from mpmath
  y2[i] = whitw(k, mw, z)
  norm = norm + y2[i] * y2[i]

norm = norm * (r_coor[2] - r_coor[1])
print('norm=',norm)
plt.title("Whittaker functions")
plt.plot(r_coor, y1, '-r', label='regular at large distances')
plt.plot(r_coor, y2, '-b', label='diverges at large distances')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.xlabel('$r$, (a.u.)')
plt.grid()
plt.show() 
```

### 3. Coulomb functions for positive enegies


The regular Coulomb wave function
    $$
    F_l(\eta,z) = C_l(\eta) z^{l+1} e^{-iz} \, {}_{1}F_{1}(l+1-i\eta, 2l+2, 2iz)
    $$

is calculated by using fonction **coulombf()**  where the normalization constant $C_l(\eta)$ is calculated by ** coulombc()**. 

This function solves the differential equation
    $$f''(z) + \left(1-\frac{2\eta}{z}-\frac{l(l+1)}{z^2}\right) f(z) = 0.$$

A second linearly independent solution is given by the irregular Coulomb wave function $G_l(\eta,z)$ (see coulombg()). 
The Coulomb wave functions with real parameters are defined in Abramowitz & Stegun, section 14. However, all parameters are permitted to be complex in this implementation (see references).

***

To solve the Schrodinger equation for the Coulomb potential
$$
-\frac{\hbar^2}{2m_e}\frac{d^2\chi}{dr^2}+\left(-E+\frac{\hbar^2 l(l+1)}{2m_er^2}-\frac{Ze^2}{r}\right)\chi=0
$$
with $E>0$ we transform it into a form where the Colomb functions (see above) can be applied. 

After the change $r=az$ and $\chi=bf$ the Schrodinger equation is transformed to
$$
\frac{\hbar^2 b}{2m_ea^2}\frac{d^2w}{dz^2}+\left(bE-\frac{\hbar^2 b l(l+1)}{2m_ea^2z^2}+\frac{Ze^2b}{az}\right)f=0
$$
with $bE=1$ ($b>0$) and $\frac{\hbar^2 b}{2ma^2}=1$ the equation is brought to the top one


$$\frac{d^2f}{dz^2} - \left(1+\frac{Ze^2}{\hbar}\sqrt{\frac{2m_e}{E}}\frac{1}{z}- \frac{l(l+1)}{z^2}\right) w = 0$$
i.e.
$
\eta=-\frac{Ze^2}{\hbar}\sqrt{m_e / (2E)}.$ It has the same form as the effective quantum number but with positive $E$. 

The scaling factor $a$ for coordinate is
$
a= \hbar/\sqrt{2m_eE}.
$


```python
from mpmath import coulombf, coulombg
import matplotlib.pyplot as plt
import numpy as np


energy_f = 1./2
l_partial_f = 1.

eta = -np.sqrt(1./(2*energy_f)) 
scale = np.sqrt(1./(2.*energy_f)) #The scaling factor $a$ for coordinate.

npoints = 1000
r_coor = np.linspace(0.1,10., npoints)
chi_f1 = np.zeros( npoints,float)
chi_f2 = np.zeros( npoints,float)
norm = 0.

for i in range(npoints):
  z = r_coor[i] / scale
  chi_f1[i] = coulombf(l_partial_f,eta, z)
  chi_f2[i] = coulombg(l_partial_f,eta, z)

plt.title("Coulomb functions")
plt.plot(r_coor,  chi_f1, '-r', label='coulomb F')
plt.plot(r_coor,  chi_f2, '-b', label='coulomb G')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid()
plt.xlabel('$r$, (a.u.)')
plt.ylabel('$\\chi_i$',size=15)
plt.ylim(-2, 2)
plt.show() 
```

## 4. Photoionization cross section calculation

```python
from mpmath import *
import matplotlib.pyplot as plt
import numpy as np
from sympy.physics.wigner import wigner_3j 
from sympy.physics.wigner import clebsch_gordan  #https://docs.sympy.org/latest/modules/physics/wigner.html#


#parameters to vary in interface boxes: 

  # atom - values: Na, Rb, Art (artifical)
  # n_i, l_partial_i, m_i, quan_def_i, -
#     'principal quantum number of the initial state',
#     'initial partial wave and its projection on the axixs of polarization of the laser light',
#     'initial quantum defect'

  # l_partial_f,    quan_def_f  - 'final partial wave', 'final quantum defect'
  # energy_max - 'maximum photon energy'
  # nE - 'number of energy points'
  # rmax - 'maximum e-atom distance (r)'
  # npoints - 'number of points along r'

#j1=1;j2=2;j3=3
#m1=0;m2=0;m3=0
#t1=clebsch_gordan(j1, j2, j3, m1, m2, m3)
#print(float(t1))
#t2=(-1)**(j1+j2-m3)*sqrt(2.*j3+1.)*wigner_3j(j1,j2,j3, m1,m2,m3)
#print(float(t2))


au2cm = 5.291772e-9
au2ev = 27.2114

# coordinate grid
npoints = 200
rmax = 25. 
r_ini = 0.5

###  Na
# initial wave function
# quan_def_i = 1.3479; n_i = 3; l_partial_i = 0; m_i = 0 # Na(S)
quan_def_i = 0.8; n_i = 3; l_partial_i = 1; m_i = 0 # Na(P)

# final wave functions
# l_partial_f = 1;    quan_def_f = 0.8833 # Na(P)
l_partial_f = 2;  quan_def_f = 0.0148972 # Na(D)

###  Rb
# initial wave function
#quan_def_i = 3.131180; n_i = 5; l_partial_i = 0; m_i = 0 # Rb(S)
#quan_def_i = 2.654884; n_i =5; l_partial_i = 1; m_i = 0 # Rb(P)

# final wave functions
#l_partial_f = 1;    quan_def_f = 2.654884 # Rb(P)
#l_partial_f = 2;    quan_def_f = 1.34809171 # Rb(D)


    
energy_i = -1./(2.* (n_i - quan_def_i)**2)

print ('initial energy is', energy_i)

# Wittaker fonctions
#################################################

k = np.sqrt(1./(-2 * energy_i))
m = np.sqrt(l_partial_i * (l_partial_i + 1) + 0.25)
scale = np.sqrt(1./(-8. * energy_i))

r_coor = np.linspace(0.5, rmax, npoints)
dr = r_coor[2] - r_coor[1]
y1 = np.zeros(npoints)
chi_i = np.zeros( npoints)
norm = 0.

for ip in range(npoints):
  z = r_coor[ip]/scale
  y1[ip] = whitm(k, m, z)
  chi_i[ip] = whitw(k, m, z)
  norm = norm + chi_i[ip]*chi_i[ip]
    

norm = norm * dr
chi_i = chi_i/np.sqrt(norm)

# plotting the initial wave function
plt.title("Initial state $\\chi_i(r)$")
plt.plot(r_coor, chi_i, '-b', label='Whittaker function regular at infinity')  # regular at infinity
plt.legend(loc='best', fancybox=True, shadow=True)
plt.xlabel('$r$, (a.u.)')
plt.ylabel('$\\chi_i$',size=15)
plt.grid()
plt.show() 


# Colomb fonctions
################################################

nE = 50
energy_max = 27. #eV
energy_max = energy_max/au2ev + energy_i # maximum scattering electron energy

chi_f = np.zeros([nE, npoints],dtype = float)
r_fi = np.zeros(nE,dtype = float)
energy_f = np.zeros(nE,dtype = float)
for i_energy in range(nE):
  energy_f[i_energy] = energy_max/nE * (i_energy + 1)
  eta = -np.sqrt(1./(2 * energy_f[i_energy]))
  scale = np.sqrt(1./(2. * energy_f[i_energy]))

  for ip in range(npoints):
    z = r_coor[ip] / scale
    coul_f = coulombf(l_partial_f,eta, z)
    coul_g = coulombg(l_partial_f,eta, z)
    
    chi_f[i_energy,ip] = coul_f * np.cos(np.pi * quan_def_f) + coul_g * np.sin(np.pi * quan_def_f)
    
    r_fi[i_energy] = r_fi[i_energy] + chi_f[i_energy,ip] * r_coor[ip] * chi_i[ip] * dr

# wave function should be energy-normalized    
  wf_normalization_factor = np.sqrt(2./(np.pi * np.sqrt(2 * energy_f[i_energy])))
  chi_f[i_energy,:] = chi_f[i_energy,:] * wf_normalization_factor
  r_fi[i_energy] = r_fi[i_energy] * wf_normalization_factor
    
# plotting one of the final wave functions
iE = nE-1
print('Energy-normalized final state for energy E=',energy_f[iE] * 27.3,'eV')
plt.title('Final state $\\chi_f(r)$')
plt.plot(r_coor, chi_f[iE-1,:], '-r', label='continuum Coulomb function')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.ylim(-2, 2)
plt.xlabel('$r$, (a.u.)')
plt.ylabel('$\\chi_f$',size=15)
plt.grid()
plt.show() 


# PI cross section
#################################################
cross_section = np.zeros(nE,dtype = float)
cross_section2 = np.zeros(nE,dtype = float)

# averaged over initial polarizations
for i_energy in range(nE):
  cross_section[i_energy] = (clebsch_gordan(l_partial_i, 1, l_partial_f, 0, 0, 0))**2  \
    *abs(r_fi[i_energy])**2 /3. * 4*np.pi**2 /137.*(energy_f[i_energy] - energy_i)

# PI cross section for initially polarized atoms
for i_energy in range(nE):
  cross_section2[i_energy] = (clebsch_gordan(l_partial_i, 1, l_partial_f, m_i, 0, m_i) *\
                            clebsch_gordan(l_partial_i, 1, l_partial_f, 0, 0, 0))**2  \
    *abs(r_fi[i_energy])**2 /3. * 4*np.pi**2 /137. * (energy_f[i_energy]-energy_i)

energy_f = (energy_f-energy_i) * au2ev
cross_section = cross_section * (au2cm)**2
cross_section2 = cross_section2 * (au2cm)**2

print('Photoinization cross section')
plt.plot(energy_f, cross_section, '-r', label='unpolarized')
plt.plot(energy_f, cross_section2,'-b', label='polarized')
plt.yscale("log");plt.ylabel('Cross section, (cm$^2$)');#plt.ylim((1.e-5,0.1))
plt.xlabel('Photon energy, (eV)')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid()
plt.show() 
```

### 3. Application

One application of gas photoionization is the measurement of small concentrations (ranging from sub-parts per billion to 10,000 parts per million (ppm)) of volatile organic compounds (VOCs) and other gases using photoionization detectors (PIDs), as illustrated in Figure 1.

The PID is a portable, nonspecific, vapor/gas detector employing the principle of photoionization to detect a variety of chemical compounds, both organic and inorganic, in air.

<img src="figures/Fig1_PID.png" alt="PID_scheme" width="30%"  />

between the electrodes.

</div>

The operation of a Photoionization Detector (PID) relies on measuring the current generated by the ionization of gases and vapors when exposed to photons emitted by an ultraviolet (UV) light source (1). The energy of these photons is typically falls between 8.4 eV and 11.7 eV, depending on the type of lamp used. These energy levels determine which compounds a PID can detect. The UV light source emits radiation into an ionization chamber (3) that contains two electrodes (2). One electrode is connected to a power source, and the other to an electrometer.

When a gas sample is introduced into the ionization chamber, components with ionization energies lower than the photon energy of the lamp are ionized (for example, a 10.6 eV lamp can detect compounds like benzene with ionization potential around 9.25 eV). This ionization process generates a current within the chamber, and the current's magnitude is directly proportional to the concentration of ionizable impurities in the sample. Notably, components of clean air, such as oxygen, nitrogen, argon, methane, carbon monoxide, and carbon dioxide, are not ionized under these conditions, as their ionization energies typically exceed 12.0 eV.

As a result, the PID is a highly efficient and cost-effective tool for monitoring potential worker exposure to VOCs, such as solvents, fuels, degreasers, plastics, and other compounds during manufacturing processes.

Advantages of PIDs are high sensitivity, fast response time, stability and durability, long service life. However, PIDs have limitations, such as non-selectivity and sensitivity to humidity. Despite these factors, PIDs remain highly effective as a rapid method for measuring VOC concentrations in air, within a wide temperature range of -30C to 50C.

### *Determining Gas Concentration from Photoionization Current*

A Photoionization Detector (PID) is used to measure the concentration of **benzene** in a room under standart condition ($P$ = 101325 Pa and $T$ = 298 K). The detector ionizes gas molecules using an ultraviolet lamp with photon energy 10.6eV. Calculate the current generated in the detector if the benzene concentration detected is 3.6 ppm. 
For calculation use the following parameters:
 - Charge of an electron ($e$): $1.602 \times 10^{19}$ C;
 - Ionization cross-section of benzene: determine it from an [article](Bibliographie/PhionozationCS_bensene.pdf).

The *current* in the PID is determined by the rate of ion production, which depends on:

- Concentration of benzene: How many benzene molecules are present in the detector's gas stream.
- Photon flux from the UV lamp: How many photons are available to ionize benzene molecules.
- Ionization cross-section: The likelihood of a benzene molecule being ionized when hit by a photon.

Once ions are produced, they are collected at the electrodes. The current is proportional to the number of ions collected per second:

$$I = e \cdot N_{i}$$

where $N_{i}$ is the number of ions prodused per second which is proportional to the concentration of the gas.

$$N_{i} = N_{m}  \sigma  \Phi_{ph}$$

where  
$\sigma$ - photoionization cross-section [$m^2$], determin from [T.A. Cool et al. in Fig.5](Bibliographie/PhionozationCS_bensene.pdf). 
For energy photons 10,6 eV, $\sigma$ = 29 Mb = $29 \times 10^{-22} m^{2}$.

For calculation of the number of detected molecules $N_{m}$ into the ionization chamber use the equetion:

$$ N_{m} = C_{gas} \times N = C_{gas} \times n \times V $$

where 

$C_{gas}$ - consentration of detected molecules (look at the problem statement).

$N$, $n$ - a number and consentration of all molecules into the ionization chamber, respectively. 

$V$ - volume of the ionization chambre (can be obtained from the technical documentation of teh PI detector).

We can get $n$ from the [Ideal gas law](https://en.wikipedia.org/wiki/Ideal_gas_law): $p = nk_BT$ ,

where $k_B =  1.380649  10^{23} JK^{-1}$ is The Boltzmann constant. So,

$$ N_m = C_{gas} \frac{p}{k_B T} V $$

$\Phi_{ph}$ - photon flux represents the number of photons emitted by the lamp per second [photons/s/cm^2
]. In the context of a PID, it determines how many UV photons are available to ionize gas molecules in the detector chamber.

$$ Phi = \frac {P_{lamp}}{E} $$
 

 $P_{lamp}$- power of the lamp [J/s];

 
 $E$ - energy of a single photon [J];

```python
#Benzene
#Ionisation potential for benzene = 9.4eV 
C_ppm = 36  #[ppm] 
e =  1.602e-19      #[C]
sigma = 29e-22     #[m2]

P = 20e-3           #[W]
E = 10.6 * e        #[J]

V = 230 * 74 * 55 * 10e-9
T = 300
p = 101325          #[Pa]
k = 1.380649 * 10e-23 # [J/K]
################################################################################

#let's find the concentration value in [1/m3]
C = (C_ppm / 1e6)
print ('C = ', C, '1/m3') 

# find photon flux$P$ = 101325 Pa
Phi = P / E
print ('Phi = ', Phi, '1/s') 

#  the number of molecules benzene in the 
N = C * p / (k * T) * V

# Current
I = e * N * sigma * Phi

print ('I = ', I, 'A') 
```
